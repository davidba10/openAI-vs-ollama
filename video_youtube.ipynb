```python
#Prueba 1
"""
An√°lisis exhaustivo: Prueba 1 - Generaci√≥n de C√≥digo
GPT-4o vs Qwen 2.5 Coder vs DeepSeek Coder
Muestra c√≥digo + an√°lisis completo
"""

import json
import ast
import sys
from typing import Dict, Any

# Colores
class Colors:
    HEADER = '\033[95m'
    BLUE = '\033[94m'
    CYAN = '\033[96m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

def print_section(title: str):
    print(f"\n{Colors.BOLD}{Colors.HEADER}{'='*80}{Colors.ENDC}")
    print(f"{Colors.BOLD}{Colors.HEADER}{title:^80}{Colors.ENDC}")
    print(f"{Colors.BOLD}{Colors.HEADER}{'='*80}{Colors.ENDC}\n")

def print_code_section(model_name: str, code: str, color: str):
    """Imprime c√≥digo con formato bonito"""
    print(f"\n{Colors.BOLD}{color}{'‚îÄ'*80}{Colors.ENDC}")
    print(f"{Colors.BOLD}{color}  {model_name}{Colors.ENDC}")
    print(f"{Colors.BOLD}{color}{'‚îÄ'*80}{Colors.ENDC}\n")
    
    # Numerar l√≠neas
    lines = code.split('\\n')
    for i, line in enumerate(lines, 1):
        # Resaltar palabras clave
        highlighted = line
        
        # Docstring
        if '"""' in line or "'''" in line:
            highlighted = f"{Colors.CYAN}{line}{Colors.ENDC}"
        # Type hints
        elif '->' in line or ': tuple' in line or ': float' in line:
            highlighted = f"{Colors.YELLOW}{line}{Colors.ENDC}"
        # Error handling
        elif 'raise' in line or 'except' in line or 'try:' in line:
            highlighted = f"{Colors.RED}{line}{Colors.ENDC}"
        # Def
        elif 'def ' in line:
            highlighted = f"{Colors.BOLD}{Colors.GREEN}{line}{Colors.ENDC}"
        else:
            highlighted = line
        
        print(f"{Colors.BOLD}{i:3}{Colors.ENDC} ‚îÇ {highlighted}")
    
    print(f"{Colors.BOLD}{color}{'‚îÄ'*80}{Colors.ENDC}\n")

def print_result(label: str, value: Any, passed: bool = None, indent: int = 0):
    """Imprime resultado con formato"""
    spaces = "  " * indent
    if passed is None:
        print(f"{spaces}{Colors.BOLD}{label}:{Colors.ENDC} {value}")
    elif passed:
        print(f"{spaces}{Colors.GREEN}‚úÖ {label}:{Colors.ENDC} {value}")
    else:
        print(f"{spaces}{Colors.RED}‚ùå {label}:{Colors.ENDC} {value}")

# ==================== DATOS REALES ====================

GPT4O_JSON = """{
  "code": "def euclidean_distance(point1: tuple, point2: tuple) -> float:\\n    \\"\\"\\"\\n    Calcula la distancia euclidiana entre dos puntos en un plano 2D.\\n\\n    Par√°metros:\\n    point1 (tuple): Una tupla que contiene dos valores num√©ricos, las coordenadas x e y del primer punto.\\n    point2 (tuple): Una tupla que contiene dos valores num√©ricos, las coordenadas x e y del segundo punto.\\n\\n    Retorna:\\n    float: La distancia euclidiana entre los dos puntos.\\n    \\"\\"\\"\\n    if not (isinstance(point1, tuple) and isinstance(point2, tuple)):\\n        raise TypeError('Los inputs deben ser tuplas.')\\n    if not (len(point1) == 2 and len(point2) == 2):\\n        raise ValueError('Las tuplas deben contener exactamente dos elementos.')\\n    try:\\n        x1, y1 = point1\\n        x2, y2 = point2\\n        return ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5\\n    except TypeError:\\n        raise TypeError('Los elementos de las tuplas deben ser num√©ricos.')",
  "explanation": "La funci√≥n euclidean_distance calcula la distancia euclidiana entre dos puntos dados en un plano 2D. Verifica que los inputs sean tuplas de dos elementos num√©ricos y lanza errores si no es as√≠, luego aplica la f√≥rmula para obtener la distancia.",
  "features": {
    "has_docstring": true,
    "has_type_hints": true,
    "has_error_handling": true,
    "function_name": "euclidean_distance"
  },
  "complexity": "b√°sica",
  "estimated_lines": 17
}"""

QWEN_JSON = """{
  "code": "def euclidean_distance(point1: tuple, point2: tuple) -> float:\\n    \\"\\"\\"Calcula la distancia euclidiana entre dos puntos en un plano 2D.\\n\\n    :param point1: Tupla con las coordenadas (x, y) del primer punto\\n    :param point2: Tupla con las coordenadas (x, y) del segundo punto\\n    :return: Distancia euclidiana entre los dos puntos como un float\\"\\"\\"\\n    if not isinstance(point1, tuple) or len(point1) != 2:\\n        raise ValueError('El primer punto debe ser una tupla con dos elementos')\\n    if not isinstance(point2, tuple) or len(point2) != 2:\\n        raise ValueError('El segundo punto debe ser una tupla con dos elementos')\\n    x1, y1 = point1\\n    x2, y2 = point2\\n    return ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5",
  "explanation": "Esta funci√≥n calcula la distancia euclidiana entre dos puntos en un plano 2D. Acepta dos tuplas como par√°metros, cada una con las coordenadas x e y del punto. Realiza comprobaciones de entrada para asegurarse de que los par√°metros son v√°lidos y retorna el resultado como un float.",
  "features": {
    "has_docstring": true,
    "has_type_hints": true,
    "has_error_handling": true,
    "function_name": "euclidean_distance"
  },
  "complexity": "b√°sica",
  "estimated_lines": 8
}"""

DEEPSEEK_JSON = """{
  "code": "def euclidean_distance(point1: Tuple[float, float], point2: Tuple[float, float]) -> float: \\n    \\"\\"\\"\\n     Calcula la distancia Euclidiana entre dos puntos en un plano bidimensional. \\n     Par√°metros: \\n         point1 (tupla con coordenadas x e y)\\n         point2 (tupla con coordenadas x e y)\\n     Devuelve:\\n         distancia calculada como float\\n    \\"\\"\\"\\n   \\n    if not(isinstance(point1, tuple) and isinstance(point2, tuple)): \\n        raise ValueError(\\"Ambos puntos deben ser tuplas de longitud 2.\\") \\n    elif len(point1) != 2 or len(point2) != 2:\\n         raise ValueError(\\"Ambos puntos deben ser tuplas de longitud 2.\\") \\n   \\n    x1, y1 = point1\\n    x2, y2 = point2\\n   \\n    return (((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5)",
  "explanation": "Esta funci√≥n implementa la f√≥rmula de distancia euclidiana para calcular la distancia entre dos puntos en un plano bidimensional.",
  "features": {
    "has_docstring": true,
    "has_type_hints": true,
    "has_error_handling": true,
    "function_name": "euclidean_distance"
  },
  "complexity": "b√°sica",
  "estimated_lines": 13
}"""

# Informaci√≥n de modelos
MODEL_INFO = {
    'gpt4o': {
        'name': 'GPT-4o',
        'params': '1-2 Trillones',
        'cost_per_1k': 0.08,  # Estimado por request
        'speed': 5.2,  # segundos promedio
        'color': Colors.RED
    },
    'qwen': {
        'name': 'Qwen 2.5 Coder 7B',
        'params': '7 Billones',
        'cost_per_1k': 0.00,
        'speed': 3.1,  # segundos promedio en GPU
        'color': Colors.GREEN
    },
    'deepseek': {
        'name': 'DeepSeek Coder 6.7B',
        'params': '6.7 Billones',
        'cost_per_1k': 0.00,
        'speed': 2.8,  # segundos promedio en GPU
        'color': Colors.BLUE
    }
}

# ==================== EVALUADOR ====================

class CodeEvaluator:
    def __init__(self, model_key: str, json_data: str):
        self.model_key = model_key
        self.info = MODEL_INFO[model_key]
        self.data = json.loads(json_data)
        self.score = 0
        self.max_score = 0
        self.results = {}
        
    def evaluate_all(self) -> Dict[str, Any]:
        """Ejecuta todas las evaluaciones"""
        
        # 1. Sintaxis v√°lida (10 pts)
        self.evaluate_syntax()
        
        # 2. Features requeridas (40 pts - 10 cada una)
        self.evaluate_features()
        
        # 3. Calidad del docstring (15 pts)
        self.evaluate_docstring_quality()
        
        # 4. Manejo de errores (20 pts)
        self.evaluate_error_handling()
        
        # 5. Eficiencia del c√≥digo (10 pts)
        self.evaluate_efficiency()
        
        # 6. Legibilidad (5 pts)
        self.evaluate_readability()
        
        return {
            'model_key': self.model_key,
            'model_name': self.info['name'],
            'score': self.score,
            'max_score': self.max_score,
            'percentage': (self.score / self.max_score * 100) if self.max_score > 0 else 0,
            'code': self.data['code'],
            'lines': self.data['estimated_lines'],
            'results': self.results
        }
    
    def evaluate_syntax(self):
        """Test 1: ¬øSintaxis v√°lida?"""
        self.max_score += 10
        try:
            # Limpiar y parsear c√≥digo
            code = self.data['code'].replace('\\n', '\n').replace('\\"', '"')
            ast.parse(code)
            self.score += 10
            self.results['syntax'] = True
        except SyntaxError:
            self.results['syntax'] = False
    
    def evaluate_features(self):
        """Test 2: Features requeridas"""
        self.max_score += 40
        features = self.data['features']
        
        points = 0
        if features.get('has_docstring'):
            points += 10
            self.results['docstring'] = True
        else:
            self.results['docstring'] = False
        
        if features.get('has_type_hints'):
            points += 10
            self.results['type_hints'] = True
        else:
            self.results['type_hints'] = False
        
        if features.get('has_error_handling'):
            points += 10
            self.results['error_handling'] = True
        else:
            self.results['error_handling'] = False
        
        if features.get('function_name') == 'euclidean_distance':
            points += 10
            self.results['correct_name'] = True
        else:
            self.results['correct_name'] = False
        
        self.score += points
    
    def evaluate_docstring_quality(self):
        """Test 3: Calidad del docstring"""
        self.max_score += 15
        code = self.data['code']
        
        points = 0
        # Longitud adecuada (5 pts)
        if '"""' in code:
            docstring_content = code.split('"""')[1] if '"""' in code else ""
            if len(docstring_content) > 50:
                points += 5
        
        # Documenta par√°metros (5 pts)
        if ':param' in code or 'Par√°metros:' in code or 'point1' in code:
            points += 5
        
        # Documenta retorno (5 pts)
        if ':return' in code or 'Retorna:' in code or 'Devuelve:' in code:
            points += 5
        
        self.score += points
        self.results['docstring_quality'] = points
    
    def evaluate_error_handling(self):
        """Test 4: Calidad del error handling"""
        self.max_score += 20
        code = self.data['code']
        
        points = 0
        # Tiene error handling (10 pts)
        if 'raise' in code:
            points += 10
        
        # M√∫ltiples validaciones (5 pts)
        raise_count = code.count('raise')
        if raise_count >= 2:
            points += 5
        
        # Valida tipos (5 pts)
        if 'isinstance' in code:
            points += 5
        
        self.score += points
        self.results['error_handling_quality'] = points
    
    def evaluate_efficiency(self):
        """Test 5: Eficiencia del c√≥digo"""
        self.max_score += 10
        code = self.data['code']
        lines = self.data['estimated_lines']
        
        points = 0
        # No usa imports innecesarios dentro de funci√≥n (5 pts)
        if 'import' not in code or code.index('import') < code.index('def'):
            points += 5
        
        # Longitud razonable (5 pts)
        if 10 <= lines <= 25:
            points += 5
        elif lines < 10:
            points += 3  # Muy corto, podr√≠a faltar validaci√≥n
        else:
            points += 2  # Muy largo
        
        self.score += points
        self.results['efficiency'] = points
    
    def evaluate_readability(self):
        """Test 6: Legibilidad"""
        self.max_score += 5
        code = self.data['code']
        
        points = 0
        # Variables con nombres claros (3 pts)
        if 'x1' in code and 'y1' in code and 'x2' in code and 'y2' in code:
            points += 3
        
        # Usa ** 0.5 o math.sqrt (2 pts)
        if '** 0.5' in code or 'math.sqrt' in code:
            points += 2
        
        self.score += points
        self.results['readability'] = points

# ==================== FUNCI√ìN PRINCIPAL ====================

def main():
    print(f"{Colors.BOLD}{Colors.CYAN}")
    print("""
    ‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
    ‚ïë                                                                   ‚ïë
    ‚ïë         ü•ä PRUEBA 1: GENERACI√ìN DE C√ìDIGO PYTHON ü•ä              ‚ïë
    ‚ïë                                                                   ‚ïë
    ‚ïë                  Distancia Euclidiana 2D                          ‚ïë
    ‚ïë                                                                   ‚ïë
    ‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù
    """)
    print(Colors.ENDC)
    
    # Mostrar tarea
    print(f"{Colors.YELLOW}üìã TAREA:{Colors.ENDC}")
    print("   Crear funci√≥n que calcule distancia euclidiana entre dos puntos")
    print("   F√≥rmula: distancia = ‚àö[(x2 - x1)¬≤ + (y2 - y1)¬≤]")
    print("\n" + f"{Colors.YELLOW}‚úÖ REQUISITOS:{Colors.ENDC}")
    print("   ‚Ä¢ Nombre: euclidean_distance")
    print("   ‚Ä¢ Type hints completos")
    print("   ‚Ä¢ Docstring detallado")
    print("   ‚Ä¢ Error handling robusto")
    
    # ==================== MOSTRAR C√ìDIGOS ====================
    
    print_section("üìù C√ìDIGO GENERADO POR CADA MODELO")
    
    # GPT-4o
    gpt4o_data = json.loads(GPT4O_JSON)
    print_code_section(
        "üî¥ GPT-4o (1-2 Trillones par√°metros | $0.08/request)",
        gpt4o_data['code'],
        Colors.RED
    )
    
    # Qwen
    qwen_data = json.loads(QWEN_JSON)
    print_code_section(
        "üü¢ Qwen 2.5 Coder 7B (7 Billones par√°metros | $0.00)",
        qwen_data['code'],
        Colors.GREEN
    )
    
    # DeepSeek
    deepseek_data = json.loads(DEEPSEEK_JSON)
    print_code_section(
        "üîµ DeepSeek Coder 6.7B (6.7 Billones par√°metros | $0.00)",
        deepseek_data['code'],
        Colors.BLUE
    )
    
    # ==================== AN√ÅLISIS COMPARATIVO ====================
    
    print_section("üîç AN√ÅLISIS COMPARATIVO VISUAL")
    
    print(f"{Colors.BOLD}{'Caracter√≠stica':<30} {'GPT-4o':<15} {'Qwen':<15} {'DeepSeek':<15}{Colors.ENDC}")
    print("‚îÄ" * 80)
    
    comparisons = [
        ("Longitud (l√≠neas)", 
         str(gpt4o_data['estimated_lines']), 
         str(qwen_data['estimated_lines']), 
         str(deepseek_data['estimated_lines'])),
        
        ("Docstring", 
         "‚úÖ Detallado", 
         "‚úÖ Sphinx", 
         "‚úÖ Est√°ndar"),
        
        ("Type Hints", 
         "‚úÖ tuple", 
         "‚úÖ tuple", 
         "‚úÖ Tuple[float,float]"),
        
        ("Error Handling", 
         "‚úÖ 3 validaciones", 
         "‚úÖ 2 validaciones", 
         "‚úÖ 2 validaciones"),
        
        ("Validaci√≥n Tipos", 
         "‚úÖ isinstance", 
         "‚úÖ isinstance", 
         "‚úÖ isinstance"),
        
        ("Validaci√≥n Longitud", 
         "‚úÖ len()", 
         "‚úÖ len()", 
         "‚úÖ len()"),
        
        ("Imports", 
         "‚úÖ Ninguno", 
         "‚úÖ Ninguno", 
         "‚ö†Ô∏è  Usa Tuple (no importado)"),
        
        ("Espacios/Formato", 
         "‚úÖ Limpio", 
         "‚úÖ Muy limpio", 
         "‚ö†Ô∏è  Espacios extra"),
    ]
    
    for feature, gpt_val, qwen_val, deep_val in comparisons:
        print(f"{feature:<30} {gpt_val:<15} {qwen_val:<15} {deep_val:<15}")
    
    # ==================== EVALUACI√ìN AUTOM√ÅTICA ====================
    
    print_section("ü§ñ EVALUACI√ìN AUTOM√ÅTICA")
    
    evaluators = [
        CodeEvaluator('gpt4o', GPT4O_JSON),
        CodeEvaluator('qwen', QWEN_JSON),
        CodeEvaluator('deepseek', DEEPSEEK_JSON)
    ]
    
    results = []
    for evaluator in evaluators:
        result = evaluator.evaluate_all()
        results.append(result)
        
        # Mostrar resultados individuales
        info = MODEL_INFO[result['model_key']]
        print(f"\n{Colors.BOLD}{info['color']}{result['model_name']}:{Colors.ENDC}")
        print(f"  Puntuaci√≥n: {result['score']}/{result['max_score']} ({result['percentage']:.1f}%)")
        print(f"  L√≠neas: {result['lines']}")
        print(f"  Velocidad promedio: {info['speed']}s")
        print(f"  Costo: ${info['cost_per_1k']:.2f}/request")
    
    # Ordenar por puntuaci√≥n
    results.sort(key=lambda x: x['score'], reverse=True)
    
    # ==================== RESULTADOS FINALES ====================
    
    print_section("üèÜ RESULTADOS FINALES")
    
    print(f"{Colors.BOLD}{'Modelo':<30} {'Score':<12} {'Calidad':<12} {'Estado'}{Colors.ENDC}")
    print("‚îÄ" * 80)
    
    for i, result in enumerate(results):
        medal = ["ü•á", "ü•à", "ü•â"][i]
        info = MODEL_INFO[result['model_key']]
        percentage = result['percentage']
        bar_length = int(percentage / 2)
        bar = "‚ñà" * bar_length + "‚ñë" * (50 - bar_length)
        
        print(f"{info['color']}{medal} {result['model_name']:<27}{Colors.ENDC} "
              f"{result['score']}/{result['max_score']:<8} "
              f"{percentage:.1f}%{' '*6} "
              f"{info['color']}{bar}{Colors.ENDC}")
    
    # ==================== AN√ÅLISIS DE COSTO/BENEFICIO ====================
    
    print_section("üí∞ AN√ÅLISIS COSTO-BENEFICIO")
    
    winner = results[0]
    winner_info = MODEL_INFO[winner['model_key']]
    
    print(f"{Colors.BOLD}Escenario: 1,000 requests/mes{Colors.ENDC}\n")
    
    for result in results:
        info = MODEL_INFO[result['model_key']]
        monthly_cost = info['cost_per_1k'] * 1
        annual_cost = monthly_cost * 12
        quality_ratio = result['percentage']
        
        print(f"{info['color']}{info['name']}:{Colors.ENDC}")
        print(f"  Calidad: {quality_ratio:.1f}%")
        print(f"  Velocidad: {info['speed']}s/request")
        print(f"  Costo mensual: ${monthly_cost:.2f}")
        print(f"  Costo anual: ${annual_cost:.2f}")
        print(f"  Ratio calidad/costo: {quality_ratio/max(monthly_cost, 0.01):.0f}")
        print()
    
    # ==================== VEREDICTO FINAL ====================
    
    print_section("üí° VEREDICTO FINAL")
    
    winner = results[0]
    winner_info = MODEL_INFO[winner['model_key']]
    
    print(f"{Colors.BOLD}{winner_info['color']}üèÜ GANADOR: {winner['model_name']}{Colors.ENDC}\n")
    print(f"  Puntuaci√≥n: {winner['score']}/{winner['max_score']} ({winner['percentage']:.1f}%)")
    print(f"  Velocidad: {winner_info['speed']}s")
    print(f"  Costo: ${winner_info['cost_per_1k']:.2f}/request")
    print(f"  Tama√±o: {winner_info['params']} par√°metros")
    
    print(f"\n{Colors.CYAN}¬øPor qu√© gan√≥?{Colors.ENDC}")
    
    # An√°lisis espec√≠fico basado en qui√©n gan√≥
    if winner['model_key'] == 'gpt4o':
        print(f"  ‚Ä¢ Docstring m√°s completo y detallado")
        print(f"  ‚Ä¢ Validaci√≥n m√°s exhaustiva (3 checks)")
        print(f"  ‚Ä¢ Mensajes de error muy descriptivos")
        print(f"  ‚Ä¢ PERO: 150-300x m√°s grande y cuesta dinero")
    elif winner['model_key'] == 'qwen':
        print(f"  ‚Ä¢ C√≥digo M√ÅS LIMPIO y conciso")
        print(f"  ‚Ä¢ Docstring estilo Sphinx (est√°ndar)")
        print(f"  ‚Ä¢ Validaci√≥n efectiva sin ser verbose")
        print(f"  ‚Ä¢ Y es GRATIS y m√°s r√°pido")
    else:
        print(f"  ‚Ä¢ Type hints m√°s espec√≠ficos (Tuple[float, float])")
        print(f"  ‚Ä¢ Validaci√≥n robusta")
        print(f"  ‚Ä¢ C√≥digo funcional")
        print(f"  ‚Ä¢ PERO: Tiene espacios extra y falta import")
    
    # Comparaci√≥n de eficiencia
    print(f"\n{Colors.YELLOW}Eficiencia comparada:{Colors.ENDC}")
    gpt_info = MODEL_INFO['gpt4o']
    qwen_info = MODEL_INFO['qwen']
    
    if results[0]['model_key'] == 'qwen' or results[0]['model_key'] == 'deepseek':
        gpt_result = next(r for r in results if r['model_key'] == 'gpt4o')
        local_result = results[0]
        
        quality_diff = gpt_result['percentage'] - local_result['percentage']
        speed_improvement = ((gpt_info['speed'] - winner_info['speed']) / gpt_info['speed']) * 100
        
        print(f"  ‚Ä¢ Diferencia de calidad: {quality_diff:.1f} puntos")
        print(f"  ‚Ä¢ Mejora de velocidad: {speed_improvement:.1f}% m√°s r√°pido")
        print(f"  ‚Ä¢ Ahorro anual: ${gpt_info['cost_per_1k'] * 12:.2f}")
        print(f"\n  {Colors.GREEN}Conclusi√≥n: El modelo local ofrece {local_result['percentage']:.0f}% de la")
        print(f"  calidad de GPT-4o, es {abs(speed_improvement):.0f}% m√°s r√°pido, y cuesta $0.{Colors.ENDC}")
    
    # Tabla resumen final
    print(f"\n{Colors.BOLD}RESUMEN EJECUTIVO:{Colors.ENDC}")
    print("‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¨‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¨‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¨‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê")
    print("‚îÇ Criterio           ‚îÇ GPT-4o   ‚îÇ Qwen     ‚îÇ DeepSeek ‚îÇ")
    print("‚îú‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î§")
    
    scores = {r['model_key']: r['percentage'] for r in results}
    print(f"‚îÇ Calidad C√≥digo     ‚îÇ {scores['gpt4o']:6.1f}%  ‚îÇ {scores['qwen']:6.1f}%  ‚îÇ {scores['deepseek']:6.1f}%  ‚îÇ")
    print(f"‚îÇ Velocidad          ‚îÇ {MODEL_INFO['gpt4o']['speed']:6.1f}s  ‚îÇ {MODEL_INFO['qwen']['speed']:6.1f}s  ‚îÇ {MODEL_INFO['deepseek']['speed']:6.1f}s  ‚îÇ")
    print(f"‚îÇ Costo/1K requests  ‚îÇ ${MODEL_INFO['gpt4o']['cost_per_1k']:7.2f} ‚îÇ ${MODEL_INFO['qwen']['cost_per_1k']:7.2f} ‚îÇ ${MODEL_INFO['deepseek']['cost_per_1k']:7.2f} ‚îÇ")
    print(f"‚îÇ Tama√±o (params)    ‚îÇ 1-2T     ‚îÇ 7B       ‚îÇ 6.7B     ‚îÇ")
    print("‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¥‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¥‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¥‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò")
    
    print(f"\n{Colors.BOLD}{Colors.GREEN}Recomendaci√≥n:{Colors.ENDC}")
    if results[0]['model_key'] != 'gpt4o':
        print(f"  Para esta tarea, {results[0]['model_name']} ofrece mejor relaci√≥n")
        print(f"  calidad/precio/velocidad. A menos que necesites el 5-10% extra de")
        print(f"  calidad que ofrece GPT-4o, los modelos locales son la mejor opci√≥n.")
    else:
        print(f"  GPT-4o ofrece la m√°xima calidad, pero los modelos locales est√°n")
        print(f"  muy cerca ({results[1]['percentage']:.0f}%) y son gratis. Para producci√≥n")
        print(f"  con volumen alto, considera h√≠brido: local para lo simple, GPT-4o")
        print(f"  para lo cr√≠tico.")
    
    print()

if __name__ == "__main__":
    main()
```

```python
#Prueba 2
"""
An√°lisis exhaustivo: Prueba 2 - An√°lisis de Sentimiento
GPT-4o vs Llama 3.1 8B vs Mistral 7B Instruct
Versi√≥n con detecci√≥n de empates
"""

import json
import sys
from typing import Dict, Any, List

# Colores
class Colors:
    HEADER = '\033[95m'
    BLUE = '\033[94m'
    CYAN = '\033[96m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

def print_section(title: str):
    print(f"\n{Colors.BOLD}{Colors.HEADER}{'='*90}{Colors.ENDC}")
    print(f"{Colors.BOLD}{Colors.HEADER}{title:^90}{Colors.ENDC}")
    print(f"{Colors.BOLD}{Colors.HEADER}{'='*90}{Colors.ENDC}\n")

def print_tweet_header(tweet_id: int, tweet_text: str, expected: str):
    """Imprime header de tweet con clasificaci√≥n esperada"""
    emoji = {"positivo": "üòä", "negativo": "üò†", "neutral": "üòê"}
    color = {"positivo": Colors.GREEN, "negativo": Colors.RED, "neutral": Colors.YELLOW}
    
    print(f"\n{Colors.BOLD}{color[expected]}{'‚îÄ'*90}{Colors.ENDC}")
    print(f"{Colors.BOLD}{color[expected]}  {emoji[expected]} TWEET {tweet_id}: {tweet_text}{Colors.ENDC}")
    print(f"{Colors.BOLD}{color[expected]}  Clasificaci√≥n correcta: {expected.upper()}{Colors.ENDC}")
    print(f"{Colors.BOLD}{color[expected]}{'‚îÄ'*90}{Colors.ENDC}\n")

def print_classification(model_name: str, result: Dict, is_correct: bool, color: str):
    """Imprime clasificaci√≥n de un modelo"""
    status = f"{Colors.GREEN}‚úÖ CORRECTO{Colors.ENDC}" if is_correct else f"{Colors.RED}‚ùå INCORRECTO{Colors.ENDC}"
    
    print(f"{Colors.BOLD}{color}{model_name}:{Colors.ENDC} {status}")
    print(f"  Clasificaci√≥n: {result['sentimiento']}")
    print(f"  Confianza: {result['confianza']:.0%}")
    print(f"  Raz√≥n ({len(result['razon'].split())} palabras): \"{result['razon']}\"")
    print()

# ==================== DATOS REALES ====================

GPT4O_JSON = """{
  "resultados": [
    {
      "tweet_id": 1,
      "sentimiento": "positivo",
      "confianza": 0.95,
      "razon": "Usa expresiones como 'incre√≠ble' y 'super√≥ mis expectativas' que indican satisfacci√≥n"
    },
    {
      "tweet_id": 2,
      "sentimiento": "negativo",
      "confianza": 0.98,
      "razon": "La frase 'servicio al cliente fue terrible' y 'nunca m√°s compro aqu√≠' expresa insatisfacci√≥n"
    },
    {
      "tweet_id": 3,
      "sentimiento": "neutral",
      "confianza": 0.85,
      "razon": "Menciona la llegada del paquete pero no expresa emoci√≥n positiva ni negativa"
    },
    {
      "tweet_id": 4,
      "sentimiento": "neutral",
      "confianza": 0.70,
      "razon": "Palabras 'pros y contras' sugieren evaluaci√≥n equilibrada sin inclinaci√≥n clara"
    },
    {
      "tweet_id": 5,
      "sentimiento": "negativo",
      "confianza": 0.99,
      "razon": "Frases como 'ESTAFA TOTAL' y 'Eviten esta empresa' indican una experiencia muy negativa"
    }
  ],
  "resumen": {
    "positivos": 1,
    "negativos": 2,
    "neutrales": 2
  }
}"""

LLAMA_JSON = """{
  "resultados": [
    {
      "tweet_id": 1,
      "sentimiento": "positivo",
      "confianza": 0.9,
      "razon": "Uso de 'Incre√≠ble' y emoticono de celebraci√≥n"
    },
    {
      "tweet_id": 2,
      "sentimiento": "negativo",
      "confianza": 1.0,
      "razon": "Uso de 'terrible', negativa experiencia con servicio al cliente"
    },
    {
      "tweet_id": 3,
      "sentimiento": "neutral",
      "confianza": 0.8,
      "razon": "Nota informativa sin opini√≥n clara sobre el producto o servicio"
    },
    {
      "tweet_id": 4,
      "sentimiento": "neutral",
      "confianza": 0.7,
      "razon": "Menci√≥n de 'pros y contras' que indica equilibrio en la evaluaci√≥n"
    },
    {
      "tweet_id": 5,
      "sentimiento": "negativo",
      "confianza": 1.0,
      "razon": "Uso de 'ESTAFA TOTAL', negativa experiencia con producto o servicio"
    }
  ],
  "resumen": {
    "positivos": 1,
    "negativos": 2,
    "neutrales": 2
  }
}"""

MISTRAL_JSON = """{
  "resultados": [
    {
      "tweet_id": 1,
      "sentimiento": "positivo",
      "confianza": 0.95,
      "razon": "La frase 'Super√≥ mis expectativas' y el emoji de felicidad indican que se trata de un comentario positivo sobre el producto"
    },
    {
      "tweet_id": 2,
      "sentimiento": "negativo",
      "confianza": 0.9,
      "razon": "La frase 'El servicio al cliente fue terrible' indica que el usuario est√° descontento con la experiencia de atenci√≥n al cliente"
    },
    {
      "tweet_id": 3,
      "sentimiento": "neutral",
      "confianza": 0.85,
      "razon": "La frase 'El paquete lleg√≥ hoy. Todo en orden.' no contiene palabras que indiquen sentimientos positivos o negativos. Se trata simplemente de una notificaci√≥n sobre la recepci√≥n del paquete"
    },
    {
      "tweet_id": 4,
      "sentimiento": "neutral",
      "confianza": 0.75,
      "razon": "La frase 'No s√© si recomendarlo, tiene pros y contras' indica que el usuario se encuentra en duda sobre su opini√≥n del producto. Por lo tanto, no se puede clasificar como sentimiento positivo o negativo con confianza alta"
    },
    {
      "tweet_id": 5,
      "sentimiento": "negativo",
      "confianza": 0.9,
      "razon": "La frase 'ESTAFA TOTAL' y 'Eviten esta empresa' indican que el usuario ha tenido una mala experiencia con la empresa y aconseja evitarla"
    }
  ],
  "resumen": {
    "positivos": 1,
    "negativos": 2,
    "neutrales": 2
  }
}"""

# Ground truth
TWEETS = [
    {"id": 1, "text": "¬°Incre√≠ble producto! Super√≥ mis expectativas üéâ", "sentiment": "positivo"},
    {"id": 2, "text": "El servicio al cliente fue terrible, nunca m√°s compro aqu√≠", "sentiment": "negativo"},
    {"id": 3, "text": "El paquete lleg√≥ hoy. Todo en orden.", "sentiment": "neutral"},
    {"id": 4, "text": "No s√© si recomendarlo, tiene pros y contras", "sentiment": "neutral"},
    {"id": 5, "text": "ESTAFA TOTAL. Me timaron. Eviten esta empresa", "sentiment": "negativo"},
]

# Informaci√≥n de modelos
MODEL_INFO = {
    'gpt4o': {
        'name': 'GPT-4o',
        'params': '1-2 Trillones',
        'cost_per_1k': 0.05,
        'speed': 4.2,
        'color': Colors.RED
    },
    'llama': {
        'name': 'Llama 3.1 8B',
        'params': '8 Billones',
        'cost_per_1k': 0.00,
        'speed': 3.1,
        'color': Colors.GREEN
    },
    'mistral': {
        'name': 'Mistral 7B Instruct',
        'params': '7 Billones',
        'cost_per_1k': 0.00,
        'speed': 2.8,
        'color': Colors.YELLOW
    }
}

# ==================== EVALUADOR ====================

class SentimentEvaluator:
    def __init__(self, model_key: str, json_data: str):
        self.model_key = model_key
        self.info = MODEL_INFO[model_key]
        self.data = json.loads(json_data)
        self.score = 0
        self.max_score = 0
        self.results = {}
        
    def evaluate_all(self) -> Dict[str, Any]:
        """Ejecuta todas las evaluaciones"""
        
        # 1. Clasificaciones correctas (50 pts - 10 por tweet)
        accuracy_score = self.evaluate_classifications()
        
        # 2. Niveles de confianza (15 pts)
        confidence_score = self.evaluate_confidence()
        
        # 3. Calidad de razones (25 pts)
        reasoning_score = self.evaluate_reasoning()
        
        # 4. Resumen correcto (10 pts)
        summary_score = self.evaluate_summary()
        
        self.score = accuracy_score + confidence_score + reasoning_score + summary_score
        self.max_score = 100
        
        return {
            'model_key': self.model_key,
            'model_name': self.info['name'],
            'score': self.score,
            'max_score': self.max_score,
            'percentage': self.score,
            'accuracy': (accuracy_score / 50) * 100,
            'correct_count': int((accuracy_score / 50) * 5),
            'confidence_score': confidence_score,
            'reasoning_score': reasoning_score,
            'results': self.results,
            'classifications': self.data['resultados']
        }
    
    def evaluate_classifications(self) -> int:
        """Eval√∫a accuracy de clasificaciones"""
        correct = 0
        for tweet in TWEETS:
            result = next((r for r in self.data['resultados'] if r['tweet_id'] == tweet['id']), None)
            if result and result['sentimiento'] == tweet['sentiment']:
                correct += 1
        
        self.results['correct_classifications'] = correct
        return correct * 10
    
    def evaluate_confidence(self) -> int:
        """Eval√∫a apropiaci√≥n de niveles de confianza"""
        points = 0
        for result in self.data['resultados']:
            conf = result['confianza']
            tweet_id = result['tweet_id']
            
            # Casos obvios (1, 2, 5) deber√≠an tener confianza alta (>0.85)
            if tweet_id in [1, 2, 5] and conf >= 0.85:
                points += 3
            # Casos ambiguos (3, 4) pueden tener confianza media-alta (0.65-0.90)
            elif tweet_id in [3, 4] and 0.65 <= conf <= 0.90:
                points += 3
            # Si la confianza est√° en rango v√°lido al menos
            elif 0 <= conf <= 1:
                points += 1
        
        self.results['confidence_score'] = points
        return min(15, points)
    
    def evaluate_reasoning(self) -> int:
        """Eval√∫a calidad de las razones"""
        points = 0
        keywords = {
            1: ['incre√≠ble', 'super√≥', 'expectativas', 'emoji'],
            2: ['terrible', 'nunca', 'servicio'],
            3: ['lleg√≥', 'orden', 'paquete', 'factual'],
            4: ['pros', 'contras', 'no s√©', 'duda'],
            5: ['estafa', 'timaron', 'eviten']
        }
        
        for result in self.data['resultados']:
            razon = result['razon'].lower()
            tweet_id = result['tweet_id']
            word_count = len(result['razon'].split())
            
            # Longitud adecuada (5 pts por tweet)
            # Penalizar muy cortas (<8 palabras) o muy largas (>35 palabras)
            if 10 <= word_count <= 30:
                points += 3  # √ìptimo
            elif 8 <= word_count <= 35:
                points += 2  # Aceptable
            elif word_count > 5:
                points += 1  # Muy corto o muy largo pero existe
            
            # Menciona palabras clave (2 pts por tweet)
            if any(kw.lower() in razon for kw in keywords.get(tweet_id, [])):
                points += 2
        
        self.results['reasoning_score'] = points
        return min(25, points)
    
    def evaluate_summary(self) -> int:
        """Eval√∫a correcci√≥n del resumen"""
        resumen = self.data['resumen']
        if (resumen['positivos'] == 1 and 
            resumen['negativos'] == 2 and 
            resumen['neutrales'] == 2):
            self.results['summary_correct'] = True
            return 10
        self.results['summary_correct'] = False
        return 0

# ==================== FUNCI√ìN PRINCIPAL ====================

def main():
    print(f"{Colors.BOLD}{Colors.CYAN}")
    print("""
    ‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
    ‚ïë                                                                           ‚ïë
    ‚ïë              ü•ä PRUEBA 2: AN√ÅLISIS DE SENTIMIENTO ü•ä                     ‚ïë
    ‚ïë                                                                           ‚ïë
    ‚ïë                    5 Tweets - 3 Modelos - 1 Ganador                      ‚ïë
    ‚ïë                                                                           ‚ïë
    ‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù
    """)
    print(Colors.ENDC)
    
    # Parsear datos
    gpt4o_data = json.loads(GPT4O_JSON)
    llama_data = json.loads(LLAMA_JSON)
    mistral_data = json.loads(MISTRAL_JSON)
    
    # ==================== MOSTRAR CLASIFICACIONES TWEET POR TWEET ====================
    
    print_section("üì± CLASIFICACIONES DETALLADAS POR TWEET")
    
    for tweet in TWEETS:
        print_tweet_header(tweet['id'], tweet['text'], tweet['sentiment'])
        
        # Obtener clasificaciones de cada modelo
        gpt4o_result = next((r for r in gpt4o_data['resultados'] if r['tweet_id'] == tweet['id']), None)
        llama_result = next((r for r in llama_data['resultados'] if r['tweet_id'] == tweet['id']), None)
        mistral_result = next((r for r in mistral_data['resultados'] if r['tweet_id'] == tweet['id']), None)
        
        # Mostrar cada clasificaci√≥n
        if gpt4o_result:
            is_correct = gpt4o_result['sentimiento'] == tweet['sentiment']
            print_classification("üî¥ GPT-4o", gpt4o_result, is_correct, Colors.RED)
        
        if llama_result:
            is_correct = llama_result['sentimiento'] == tweet['sentiment']
            print_classification("üü¢ Llama 3.1 8B", llama_result, is_correct, Colors.GREEN)
        
        if mistral_result:
            is_correct = mistral_result['sentimiento'] == tweet['sentiment']
            print_classification("üü° Mistral 7B", mistral_result, is_correct, Colors.YELLOW)
        
        print(f"{Colors.BOLD}{'‚îÄ'*90}{Colors.ENDC}")
    
    # ==================== AN√ÅLISIS DE RAZONES ====================
    
    print_section("üí¨ AN√ÅLISIS DE CALIDAD DE RAZONES")
    
    print(f"{Colors.CYAN}Comparaci√≥n de estilos:{Colors.ENDC}\n")
    
    def analyze_reason_style(data, model_name):
        total_words = sum(len(r['razon'].split()) for r in data['resultados'])
        avg_words = total_words / 5
        
        min_words = min(len(r['razon'].split()) for r in data['resultados'])
        max_words = max(len(r['razon'].split()) for r in data['resultados'])
        
        # Contar menciones de keywords
        keywords_count = 0
        keyword_sets = {
            1: ['incre√≠ble', 'super√≥', 'expectativas', 'emoji'],
            2: ['terrible', 'nunca', 'servicio'],
            3: ['lleg√≥', 'orden', 'paquete'],
            4: ['pros', 'contras', 'duda'],
            5: ['estafa', 'timaron', 'eviten']
        }
        
        for result in data['resultados']:
            razon_lower = result['razon'].lower()
            tweet_keywords = keyword_sets.get(result['tweet_id'], [])
            if any(kw in razon_lower for kw in tweet_keywords):
                keywords_count += 1
        
        # Determinar estilo
        if avg_words < 12:
            style = "‚ö° Conciso"
        elif avg_words > 20:
            style = "üìö Explicativo"
        else:
            style = "‚öñÔ∏è  Balanceado"
        
        print(f"{Colors.BOLD}{model_name}:{Colors.ENDC}")
        print(f"  Promedio: {avg_words:.1f} palabras/raz√≥n")
        print(f"  Rango: {min_words}-{max_words} palabras")
        print(f"  Keywords mencionadas: {keywords_count}/5 tweets")
        print(f"  Estilo: {style}")
        print()
    
    analyze_reason_style(gpt4o_data, "üî¥ GPT-4o")
    analyze_reason_style(llama_data, "üü¢ Llama 3.1 8B")
    analyze_reason_style(mistral_data, "üü° Mistral 7B")
    
    # ==================== EVALUACI√ìN AUTOM√ÅTICA ====================
    
    print_section("ü§ñ EVALUACI√ìN AUTOM√ÅTICA")
    
    evaluators = [
        SentimentEvaluator('gpt4o', GPT4O_JSON),
        SentimentEvaluator('llama', LLAMA_JSON),
        SentimentEvaluator('mistral', MISTRAL_JSON)
    ]
    
    results = []
    for evaluator in evaluators:
        result = evaluator.evaluate_all()
        results.append(result)
        
        info = MODEL_INFO[result['model_key']]
        print(f"\n{Colors.BOLD}{info['color']}{result['model_name']}:{Colors.ENDC}")
        print(f"  Puntuaci√≥n total: {result['score']}/100")
        print(f"  ‚îî‚îÄ Accuracy: 50/50 ({result['accuracy']:.0f}%)")
        print(f"  ‚îî‚îÄ Confianza: {result['confidence_score']}/15")
        print(f"  ‚îî‚îÄ Razones: {result['reasoning_score']}/25")
        print(f"  ‚îî‚îÄ Resumen: 10/10")
        print(f"  Velocidad: {info['speed']}s")
        print(f"  Costo: ${info['cost_per_1k']:.2f}/1K")
    
    # Ordenar por puntuaci√≥n
    results.sort(key=lambda x: x['score'], reverse=True)
    
    # Detectar empates (diferencia <= 2 puntos)
    top_score = results[0]['score']
    tied_models = [r for r in results if top_score - r['score'] <= 2]
    
    # ==================== RESULTADOS FINALES ====================
    
    print_section("üèÜ RESULTADOS FINALES")
    
    if len(tied_models) > 1:
        print(f"{Colors.BOLD}{Colors.YELLOW}‚öñÔ∏è  EMPATE T√âCNICO DETECTADO{Colors.ENDC}")
        print(f"Diferencia de ‚â§2 puntos entre los mejores modelos\n")
    
    print(f"{Colors.BOLD}{'Posici√≥n':<12} {'Modelo':<25} {'Score':<15} {'Desglose':<30}{Colors.ENDC}")
    print("‚îÄ" * 90)
    
    for i, result in enumerate(results):
        if i < len(tied_models):
            medal = "üèÜ"  # Empate
        else:
            medal = "ü•â"
        
        info = MODEL_INFO[result['model_key']]
        
        desglose = f"50+{result['confidence_score']}+{result['reasoning_score']}+10"
        
        print(f"{medal} {i+1}¬∫{'':8} {result['model_name']:<25} {result['score']}/100{'':7} {desglose}")
    
    # ==================== AN√ÅLISIS DEL EMPATE ====================
    
    if len(tied_models) > 1:
        print_section("üîç AN√ÅLISIS DEL EMPATE")
        
        print(f"{Colors.BOLD}Modelos empatados:{Colors.ENDC}")
        for tied in tied_models:
            info = MODEL_INFO[tied['model_key']]
            print(f"  {info['color']}{tied['model_name']}: {tied['score']}/100{Colors.ENDC}")
        
        print(f"\n{Colors.CYAN}¬øPor qu√© empatan?{Colors.ENDC}\n")
        print(f"  {Colors.GREEN}‚úÖ Todos tuvieron 100% de accuracy (50/50 puntos){Colors.ENDC}")
        print(f"  {Colors.GREEN}‚úÖ Todos tuvieron resumen correcto (10/10 puntos){Colors.ENDC}")
        print()
        
        print(f"{Colors.YELLOW}Las peque√±as diferencias:{Colors.ENDC}\n")
        
        for tied in tied_models:
            info = MODEL_INFO[tied['model_key']]
            print(f"  {info['color']}{tied['model_name']}:{Colors.ENDC}")
            print(f"    Confianza: {tied['confidence_score']}/15")
            print(f"    Razones: {tied['reasoning_score']}/25")
            
            # Analizar qu√© caracteriza sus razones
            data = json.loads(GPT4O_JSON if tied['model_key'] == 'gpt4o' 
                            else LLAMA_JSON if tied['model_key'] == 'llama' 
                            else MISTRAL_JSON)
            
            avg_words = sum(len(r['razon'].split()) for r in data['resultados']) / 5
            
            if tied['model_key'] == 'gpt4o':
                print(f"    Estilo: Balanceado ({avg_words:.0f} palabras/raz√≥n)")
                print(f"    Fortaleza: Confianza bien calibrada")
            elif tied['model_key'] == 'mistral':
                print(f"    Estilo: Explicativo ({avg_words:.0f} palabras/raz√≥n)")
                print(f"    Fortaleza: Razones M√ÅS contextualizadas")
            else:
                print(f"    Estilo: Conciso ({avg_words:.0f} palabras/raz√≥n)")
                print(f"    Fortaleza: Directo al punto")
            print()
        
        print(f"{Colors.CYAN}Veredicto del empate:{Colors.ENDC}")
        print(f"  Para esta tarea, GPT-4o y Mistral son INDISTINGUIBLES")
        print(f"  en t√©rminos de capacidad. La diferencia es solo estilo:")
        print()
        print(f"  {Colors.RED}‚Ä¢ GPT-4o:{Colors.ENDC} Razones balanceadas (13-14 palabras)")
        print(f"  {Colors.YELLOW}‚Ä¢ Mistral:{Colors.ENDC} Razones m√°s extensas (22-25 palabras)")
        print()
        print(f"  Ambos explican PERFECTAMENTE el razonamiento.")
    
    # ==================== VEREDICTO FINAL ====================
    
    print_section("üí° VEREDICTO FINAL")
    
    if len(tied_models) > 1:
        # Es un empate
        print(f"{Colors.BOLD}{Colors.YELLOW}üèÜ EMPATE: {' & '.join([t['model_name'] for t in tied_models])}{Colors.ENDC}\n")
        
        gpt4o_res = next(r for r in results if r['model_key'] == 'gpt4o')
        mistral_res = next(r for r in results if r['model_key'] == 'mistral')
        llama_res = next(r for r in results if r['model_key'] == 'llama')
        
        print(f"Puntuaciones finales:")
        print(f"  üî¥ GPT-4o: {gpt4o_res['score']}/100")
        print(f"  üü° Mistral: {mistral_res['score']}/100")
        print(f"  üü¢ Llama: {llama_res['score']}/100")
        
        print(f"\n{Colors.CYAN}An√°lisis:{Colors.ENDC}")
        print(f"  Los 3 modelos clasificaron PERFECTAMENTE (100% accuracy).")
        print(f"  GPT-4o y Mistral empatan en calidad total.")
        print(f"  Llama est√° muy cerca pero sus razones son m√°s cortas.")
        
        print(f"\n{Colors.YELLOW}¬øCu√°l elegir?{Colors.ENDC}\n")
        
        print(f"  {Colors.RED}Elige GPT-4o si:{Colors.ENDC}")
        print(f"    ‚Ä¢ Prefieres explicaciones BALANCEADAS")
        print(f"    ‚Ä¢ El costo no es problema")
        print(f"    ‚Ä¢ Necesitas calibraci√≥n de confianza perfecta")
        print()
        
        print(f"  {Colors.YELLOW}Elige Mistral si:{Colors.ENDC}")
        print(f"    ‚Ä¢ Prefieres explicaciones M√ÅS DETALLADAS")
        print(f"    ‚Ä¢ Quieres el modelo M√ÅS R√ÅPIDO (2.8s)")
        print(f"    ‚Ä¢ Presupuesto = $0")
        print()
        
        print(f"  {Colors.GREEN}Elige Llama si:{Colors.ENDC}")
        print(f"    ‚Ä¢ Prefieres explicaciones CONCISAS")
        print(f"    ‚Ä¢ Presupuesto = $0")
        print(f"    ‚Ä¢ M√°xima confianza en obvios (1.0)")
        
        print(f"\n{Colors.BOLD}{Colors.GREEN}üí° RECOMENDACI√ìN:{Colors.ENDC}")
        print(f"  Para an√°lisis de sentimiento b√°sico:")
        print()
        print(f"  {Colors.YELLOW}Usa Mistral 7B.{Colors.ENDC}")
        print(f"  Mismo accuracy que GPT-4o, explicaciones igual de buenas,")
        print(f"  M√ÅS R√ÅPIDO (2.8s vs 4.2s), y GRATIS.")
        print()
        print(f"  Ahorro: ${MODEL_INFO['gpt4o']['cost_per_1k'] * 10 * 12:.2f}/a√±o")
        print(f"  (basado en 10,000 clasificaciones/mes)")
        
    else:
        # Ganador claro
        winner = results[0]
        winner_info = MODEL_INFO[winner['model_key']]
        
        print(f"{Colors.BOLD}{winner_info['color']}üèÜ GANADOR: {winner['model_name']}{Colors.ENDC}\n")
        print(f"  Puntuaci√≥n: {winner['score']}/100")
        print(f"  Accuracy: {winner['accuracy']:.0f}%")
    
    print()

if __name__ == "__main__":
    main()
```

```python
#Prueba 3
"""
An√°lisis exhaustivo: Prueba 3 - Extracci√≥n de Datos Estructurados
GPT-4o vs Qwen 2.5 14B vs Mixtral 8x7B
Con datos reales de ejecuci√≥n
"""

import json
import sys
import re

# Colores
class Colors:
    HEADER = '\033[95m'
    BLUE = '\033[94m'
    CYAN = '\033[96m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

def print_section(title: str):
    print(f"\n{Colors.BOLD}{Colors.HEADER}{'='*90}{Colors.ENDC}")
    print(f"{Colors.BOLD}{Colors.HEADER}{title:^90}{Colors.ENDC}")
    print(f"{Colors.BOLD}{Colors.HEADER}{'='*90}{Colors.ENDC}\n")

# ==================== OUTPUTS REALES ====================

GPT4O_JSON = """{
  "pedidos": [
    {
      "email_id": 1,
      "cliente": {
        "nombre": "Mar√≠a Garc√≠a",
        "empresa": "TechSolutions SL",
        "email": "maria.garcia@email.com",
        "telefono": "+34 91 123 4567",
        "cif": null
      },
      "fecha_solicitud": "2024-03-15",
      "productos": [
        {
          "descripcion": "Laptop HP ProBook 450 G10",
          "cantidad": 5,
          "categoria": "hardware"
        },
        {
          "descripcion": "Monitor Dell 27\\" 4K",
          "cantidad": 2,
          "categoria": "hardware"
        },
        {
          "descripcion": "Licencia Microsoft Office 365 Business (10 usuarios)",
          "cantidad": 1,
          "categoria": "software"
        }
      ],
      "presupuesto": {
        "minimo": null,
        "maximo": 8500,
        "moneda": "EUR"
      },
      "fecha_entrega": "2024-03-30",
      "urgencia": "alta",
      "metodo_pago": "Transferencia bancaria",
      "notas": null
    },
    {
      "email_id": 2,
      "cliente": {
        "nombre": "Juan L√≥pez",
        "empresa": "InnovaStart",
        "email": "juan.lopez@startup.io",
        "telefono": "+34 93 987 6543",
        "cif": "B98765432"
      },
      "fecha_solicitud": "2024-03-18",
      "productos": [
        {
          "descripcion": "Servidor Dell PowerEdge R740 (32GB RAM, 2TB SSD)",
          "cantidad": 2,
          "categoria": "hardware"
        },
        {
          "descripcion": "Instalaci√≥n y configuraci√≥n",
          "cantidad": 1,
          "categoria": "servicio"
        },
        {
          "descripcion": "Soporte t√©cnico por 1 a√±o",
          "cantidad": 1,
          "categoria": "servicio"
        }
      ],
      "presupuesto": {
        "minimo": 12000,
        "maximo": 15000,
        "moneda": "EUR"
      },
      "fecha_entrega": "2024-04-15",
      "urgencia": "alta",
      "metodo_pago": "50% por adelantado y 50% al finalizar",
      "notas": null
    },
    {
      "email_id": 3,
      "cliente": {
        "nombre": "Ana Mart√≠nez",
        "empresa": "Comercial del Norte SL",
        "email": "ana.martinez@pyme.es",
        "telefono": "987 654 321",
        "cif": null
      },
      "fecha_solicitud": "2024-03-20",
      "productos": [
        {
          "descripcion": "Ordenador de sobremesa (i5, 16GB RAM, 512GB SSD)",
          "cantidad": 8,
          "categoria": "hardware"
        },
        {
          "descripcion": "Monitor de 24 pulgadas",
          "cantidad": 8,
          "categoria": "hardware"
        },
        {
          "descripcion": "Teclados y ratones inal√°mbricos",
          "cantidad": 8,
          "categoria": "hardware"
        },
        {
          "descripcion": "Impresora multifunci√≥n l√°ser color",
          "cantidad": 1,
          "categoria": "hardware"
        }
      ],
      "presupuesto": {
        "minimo": 6000,
        "maximo": 6000,
        "moneda": "EUR"
      },
      "fecha_entrega": "2024-05-01",
      "urgencia": "baja",
      "metodo_pago": "Contrareembolso o tarjeta",
      "notas": null
    }
  ],
  "resumen": {
    "total_pedidos": 3,
    "presupuesto_total_min": 18000,
    "presupuesto_total_max": 29500,
    "pedidos_urgentes": 2,
    "categorias": {
      "hardware": 24,
      "software": 1,
      "servicio": 2
    }
  }
}"""

QWEN_JSON = """{
  "pedidos": [
    {
      "email_id": 1,
      "cliente": {
        "nombre": "Mar√≠a Garc√≠a",
        "empresa": "TechSolutions SL",
        "email": "maria.garcia@email.com",
        "telefono": "+34 91 123 4567",
        "cif": null
      },
      "fecha_solicitud": "2024-03-15",
      "productos": [
        {
          "descripcion": "Laptop HP ProBook 450 G10",
          "cantidad": 5,
          "categoria": "hardware"
        },
        {
          "descripcion": "Monitor Dell 27\\" 4K",
          "cantidad": 2,
          "categoria": "hardware"
        },
        {
          "descripcion": "Licencia Microsoft Office 365 Business (10 usuarios)",
          "cantidad": 1,
          "categoria": "software"
        }
      ],
      "presupuesto": {
        "minimo": null,
        "maximo": 8500,
        "moneda": "EUR"
      },
      "fecha_entrega": "2024-03-30",
      "urgencia": "alta",
      "metodo_pago": "Transferencia bancaria",
      "notas": null
    },
    {
      "email_id": 2,
      "cliente": {
        "nombre": "Juan L√≥pez (CTO)",
        "empresa": "InnovaStart",
        "email": "juan.lopez@startup.io",
        "telefono": "+34 93 987 6543",
        "cif": "B98765432"
      },
      "fecha_solicitud": "2024-03-18",
      "productos": [
        {
          "descripcion": "Servidor Dell PowerEdge R740 (32GB RAM, 2TB SSD)",
          "cantidad": 2,
          "categoria": "hardware"
        },
        {
          "descripcion": "Instalaci√≥n y configuraci√≥n incluida",
          "cantidad": 1,
          "categoria": "servicio"
        },
        {
          "descripcion": "Soporte t√©cnico por 1 a√±o",
          "cantidad": 1,
          "categoria": "servicio"
        }
      ],
      "presupuesto": {
        "minimo": 12000,
        "maximo": 15000,
        "moneda": "EUR"
      },
      "fecha_entrega": "2024-04-15",
      "urgencia": "media",
      "metodo_pago": "50% por adelantado y 50% al finalizar",
      "notas": null
    },
    {
      "email_id": 3,
      "cliente": {
        "nombre": "Ana Mart√≠nez",
        "empresa": "Comercial del Norte SL",
        "email": "ana.martinez@pyme.es",
        "telefono": "987 654 321",
        "cif": null
      },
      "fecha_solicitud": "2024-03-20",
      "productos": [
        {
          "descripcion": "Ordenador de sobremesa (i5, 16GB RAM, 512GB SSD)",
          "cantidad": 8,
          "categoria": "hardware"
        },
        {
          "descripcion": "Monitor de 24 pulgadas",
          "cantidad": 8,
          "categoria": "hardware"
        },
        {
          "descripcion": "Teclados y ratones inal√°mbricos",
          "cantidad": null,
          "categoria": "hardware"
        },
        {
          "descripcion": "Impresora multifunci√≥n l√°ser color",
          "cantidad": 1,
          "categoria": "hardware"
        }
      ],
      "presupuesto": {
        "minimo": null,
        "maximo": 6000,
        "moneda": "EUR"
      },
      "fecha_entrega": null,
      "urgencia": "baja",
      "metodo_pago": "Contrareembolso o tarjeta",
      "notas": null
    }
  ],
  "resumen": {
    "total_pedidos": 3,
    "presupuesto_total_min": 12000,
    "presupuesto_total_max": 29500,
    "pedidos_urgentes": 2,
    "categorias": {
      "hardware": 18,
      "software": 1,
      "servicio": 2
    }
  }
}"""

MIXTRAL_JSON = """{
  "pedidos": [
    {
      "email_id": 1,
      "cliente": {
        "nombre": "Mar√≠a Garc√≠a",
        "empresa": "TechSolutions SL",
        "email": "maria.garcia@email.com",
        "telefono": "+34 91 123 4567",
        "cif": null
      },
      "fecha_solicitud": "2024-03-15",
      "productos": [
        {
          "descripcion": "Laptop HP ProBook 450 G10",
          "cantidad": 5,
          "categoria": "hardware"
        },
        {
          "descripcion": "Monitor Dell 27'' 4K",
          "cantidad": 2,
          "categoria": "hardware"
        },
        {
          "descripcion": "Licencia Microsoft Office 365 Business (10 usuarios)",
          "cantidad": 1,
          "categoria": "software"
        }
      ],
      "presupuesto": {
        "minimo": null,
        "maximo": 8500,
        "moneda": "EUR"
      },
      "fecha_entrega": "2024-03-30",
      "urgencia": "alta",
      "metodo_pago": "Transferencia bancaria",
      "notas": null
    },
    {
      "email_id": 2,
      "cliente": {
        "nombre": "Juan L√≥pez",
        "empresa": "InnovaStart",
        "email": "juan.lopez@startup.io",
        "telefono": "+34 93 987 6543",
        "cif": "B98765432"
      },
      "fecha_solicitud": "2024-03-18",
      "productos": [
        {
          "descripcion": "Servidor Dell PowerEdge R740 (32GB RAM, 2TB SSD)",
          "cantidad": 2,
          "categoria": "hardware"
        },
        {
          "descripcion": "Instalaci√≥n y configuraci√≥n incluida",
          "cantidad": 1,
          "categoria": "servicio"
        },
        {
          "descripcion": "Soporte t√©cnico por 1 a√±o",
          "cantidad": 1,
          "categoria": "servicio"
        }
      ],
      "presupuesto": {
        "minimo": 12000,
        "maximo": 15000,
        "moneda": "EUR"
      },
      "fecha_entrega": "2024-04-15",
      "urgencia": "media",
      "metodo_pago": "50% por adelantado y 50% al finalizar",
      "notas": null
    },
    {
      "email_id": 3,
      "cliente": {
        "nombre": "Ana Mart√≠nez",
        "empresa": "Comercial del Norte SL",
        "email": "ana.martinez@pyme.es",
        "telefono": "987 654 321",
        "cif": null
      },
      "fecha_solicitud": "2024-03-20",
      "productos": [
        {
          "descripcion": "Ordenador de sobremesa (i5, 16GB RAM, 512GB SSD)",
          "cantidad": 8,
          "categoria": "hardware"
        },
        {
          "descripcion": "Monitor de 24 pulgadas",
          "cantidad": 8,
          "categoria": "hardware"
        },
        {
          "descripcion": "Teclado y rat√≥n inal√°mbrico",
          "cantidad": 16,
          "categoria": "hardware"
        },
        {
          "descripcion": "Impresora multifunci√≥n l√°ser color",
          "cantidad": 1,
          "categoria": "hardware"
        }
      ],
      "presupuesto": {
        "minimo": 6000,
        "maximo": null,
        "moneda": "EUR"
      },
      "fecha_entrega": "2024-05-01",
      "urgencia": "baja",
      "metodo_pago": "Contrareembolso o tarjeta",
      "notas": null
    }
  ],
  "resumen": {
    "total_pedidos": 3,
    "presupuesto_total_min": 21000,
    "presupuesto_total_max": 23500,
    "pedidos_urgentes": 1,
    "categorias": {
      "hardware": 27,
      "software": 1,
      "servicio": 2
    }
  }
}"""

# ==================== INFORMACI√ìN DE MODELOS ====================

MODEL_INFO = {
    'gpt4o': {
        'name': 'GPT-4o',
        'params': '1-2 Trillones',
        'cost_per_1k': 0.05,
        'speed': 6.0,
        'vram': 'N/A (API)',
        'color': Colors.RED
    },
    'qwen': {
        'name': 'Qwen 2.5 14B',
        'params': '14 Billones',
        'cost_per_1k': 0.00,
        'speed': 10.0,
        'vram': '9GB',
        'color': Colors.GREEN
    },
    'mixtral': {
        'name': 'Mixtral 8x7B',
        'params': '47B activos (MoE)',
        'cost_per_1k': 0.00,
        'speed': 12.0,
        'vram': '26GB (GPU+RAM)',
        'color': Colors.YELLOW
    }
}

# ==================== EVALUADOR ====================

class DataExtractionEvaluator:
    def __init__(self, model_key: str, json_data: str):
        self.model_key = model_key
        self.info = MODEL_INFO[model_key]
        try:
            self.data = json.loads(json_data)
        except Exception as e:
            print(f"{Colors.RED}Error parseando {model_key}: {e}{Colors.ENDC}")
            self.data = {"pedidos": [], "resumen": {}}
        self.score = 0
        self.max_score = 100
        self.details = {}
        
    def evaluate_all(self):
        """Ejecuta todas las evaluaciones"""
        
        # 1. Estructura correcta (10 pts)
        structure_score = self.evaluate_structure()
        
        # 2. N√∫mero de pedidos (10 pts)
        pedidos_score = self.evaluate_pedidos_count()
        
        # 3. Completitud de campos (30 pts)
        completeness_score = self.evaluate_completeness()
        
        # 4. Productos extra√≠dos (20 pts)
        productos_score = self.evaluate_productos()
        
        # 5. Resumen correcto (20 pts)
        resumen_score = self.evaluate_resumen()
        
        # 6. Formato de datos (10 pts)
        format_score = self.evaluate_format()
        
        self.score = (structure_score + pedidos_score + completeness_score + 
                     productos_score + resumen_score + format_score)
        
        return {
            'model_key': self.model_key,
            'model_name': self.info['name'],
            'score': self.score,
            'max_score': self.max_score,
            'details': self.details,
            'data': self.data
        }
    
    def evaluate_structure(self) -> int:
        """Eval√∫a si tiene la estructura correcta"""
        if ('pedidos' in self.data and isinstance(self.data['pedidos'], list) and 
            'resumen' in self.data and isinstance(self.data['resumen'], dict)):
            self.details['structure'] = f"{Colors.GREEN}‚úÖ Estructura correcta{Colors.ENDC}"
            return 10
        self.details['structure'] = f"{Colors.RED}‚ùå Estructura incorrecta{Colors.ENDC}"
        return 0
    
    def evaluate_pedidos_count(self) -> int:
        """Eval√∫a n√∫mero de pedidos"""
        if not self.data.get('pedidos'):
            self.details['pedidos_count'] = f"{Colors.RED}‚ùå No hay pedidos{Colors.ENDC}"
            return 0
        
        count = len(self.data['pedidos'])
        if count == 3:
            self.details['pedidos_count'] = f"{Colors.GREEN}‚úÖ 3 pedidos extra√≠dos{Colors.ENDC}"
            return 10
        elif count == 2:
            self.details['pedidos_count'] = f"{Colors.YELLOW}‚ö†Ô∏è  2 pedidos (falta 1){Colors.ENDC}"
            return 6
        else:
            self.details['pedidos_count'] = f"{Colors.RED}‚ùå {count} pedidos{Colors.ENDC}"
            return 0
    
    def evaluate_completeness(self) -> int:
        """Eval√∫a completitud de campos en cada pedido"""
        if not self.data.get('pedidos'):
            self.details['completeness'] = f"{Colors.RED}‚ùå Sin pedidos{Colors.ENDC}"
            return 0
        
        points = 0
        complete_count = 0
        issues = []
        
        for i, pedido in enumerate(self.data['pedidos'][:3], 1):
            pedido_issues = []
            
            if not pedido.get('cliente') or not isinstance(pedido['cliente'], dict):
                pedido_issues.append('cliente')
            if not pedido.get('productos') or not isinstance(pedido['productos'], list):
                pedido_issues.append('productos')
            if not pedido.get('presupuesto') or not isinstance(pedido['presupuesto'], dict):
                pedido_issues.append('presupuesto')
            if not pedido.get('urgencia'):
                pedido_issues.append('urgencia')
            if not pedido.get('fecha_solicitud'):
                pedido_issues.append('fecha_solicitud')
            
            if not pedido_issues:
                points += 10
                complete_count += 1
            elif len(pedido_issues) <= 2:
                points += 6
                issues.append(f"Email {i}: faltan {', '.join(pedido_issues)}")
        
        if complete_count == 3:
            self.details['completeness'] = f"{Colors.GREEN}‚úÖ 3/3 pedidos completos (30/30 pts){Colors.ENDC}"
        elif complete_count >= 2:
            self.details['completeness'] = f"{Colors.YELLOW}‚ö†Ô∏è  {complete_count}/3 completos ({points}/30 pts){Colors.ENDC}"
        else:
            self.details['completeness'] = f"{Colors.RED}‚ùå {complete_count}/3 completos ({points}/30 pts){Colors.ENDC}"
        
        return min(30, points)
    
    def evaluate_productos(self) -> int:
        """Eval√∫a productos extra√≠dos"""
        if not self.data.get('pedidos'):
            self.details['productos'] = f"{Colors.RED}‚ùå Sin productos{Colors.ENDC}"
            return 0
        
        total_lineas = 0
        cantidad_null = False
        
        for pedido in self.data['pedidos']:
            if pedido.get('productos'):
                for prod in pedido['productos']:
                    total_lineas += 1
                    if prod.get('cantidad') is None:
                        cantidad_null = True
        
        # Esperamos 9-10 l√≠neas de productos
        if total_lineas >= 9 and total_lineas <= 10:
            if not cantidad_null:
                points = 20
                self.details['productos'] = f"{Colors.GREEN}‚úÖ {total_lineas} l√≠neas extra√≠das (20/20 pts){Colors.ENDC}"
            else:
                points = 17
                self.details['productos'] = f"{Colors.YELLOW}‚ö†Ô∏è  {total_lineas} l√≠neas pero cantidad=null en alguna (17/20 pts){Colors.ENDC}"
        elif total_lineas >= 7:
            points = 15
            msg = f"{total_lineas} l√≠neas extra√≠das"
            if cantidad_null:
                msg += ", cantidad=null en alguna"
            self.details['productos'] = f"{Colors.YELLOW}‚ö†Ô∏è  {msg} (15/20 pts){Colors.ENDC}"
        else:
            points = 10
            self.details['productos'] = f"{Colors.RED}‚ùå Solo {total_lineas} l√≠neas (10/20 pts){Colors.ENDC}"
        
        return points
    
    def evaluate_resumen(self) -> int:
        """Eval√∫a el resumen"""
        if not self.data.get('resumen'):
            self.details['resumen'] = f"{Colors.RED}‚ùå Sin resumen{Colors.ENDC}"
            return 0
        
        resumen = self.data['resumen']
        points = 0
        issues = []
        
        # Total pedidos (5 pts)
        if resumen.get('total_pedidos') == 3:
            points += 5
        else:
            issues.append(f"pedidos:{resumen.get('total_pedidos',0)}‚â†3")
        
        # Presupuesto total m√≠nimo (5 pts) - Esperado: ~26,500
        min_budget = resumen.get('presupuesto_total_min', 0)
        if 25000 <= min_budget <= 27000:
            points += 5
        else:
            issues.append(f"min:{min_budget}‚â†26500")
        
        # Presupuesto total m√°ximo (5 pts) - Esperado: 29,500
        max_budget = resumen.get('presupuesto_total_max', 0)
        if 28500 <= max_budget <= 30000:
            points += 5
        else:
            issues.append(f"max:{max_budget}‚â†29500")
        
        # Pedidos urgentes (5 pts) - Esperado: 2
        urgentes = resumen.get('pedidos_urgentes', 0)
        if urgentes == 2:
            points += 5
        elif urgentes == 1:
            points += 3
            issues.append(f"urgentes:{urgentes}‚â†2")
        else:
            issues.append(f"urgentes:{urgentes}‚â†2")
        
        if points == 20:
            self.details['resumen'] = f"{Colors.GREEN}‚úÖ Resumen perfecto (20/20 pts){Colors.ENDC}"
        elif points >= 15:
            self.details['resumen'] = f"{Colors.YELLOW}‚ö†Ô∏è  Resumen casi correcto ({points}/20 pts): {', '.join(issues)}{Colors.ENDC}"
        else:
            self.details['resumen'] = f"{Colors.RED}‚ùå Resumen con errores ({points}/20 pts): {', '.join(issues)}{Colors.ENDC}"
        
        return points
    
    def evaluate_format(self) -> int:
        """Eval√∫a formato de datos"""
        if not self.data.get('pedidos') or len(self.data['pedidos']) == 0:
            self.details['format'] = f"{Colors.RED}‚ùå Sin datos para formato{Colors.ENDC}"
            return 0
        
        points = 0
        issues = []
        
        primer_pedido = self.data['pedidos'][0]
        
        # Fechas en formato ISO (3 pts)
        if primer_pedido.get('fecha_solicitud'):
            if re.match(r'^\d{4}-\d{2}-\d{2}$', str(primer_pedido['fecha_solicitud'])):
                points += 3
            else:
                issues.append("fecha no ISO")
        
        # Presupuestos son n√∫meros (4 pts)
        if primer_pedido.get('presupuesto'):
            pres = primer_pedido['presupuesto']
            valid_min = isinstance(pres.get('minimo'), (int, float)) or pres.get('minimo') is None
            valid_max = isinstance(pres.get('maximo'), (int, float)) or pres.get('maximo') is None
            if valid_min and valid_max:
                points += 4
            else:
                issues.append("presupuesto no num√©rico")
        
        # Email v√°lido (3 pts)
        if primer_pedido.get('cliente', {}).get('email'):
            if '@' in str(primer_pedido['cliente']['email']):
                points += 3
            else:
                issues.append("email inv√°lido")
        
        if points >= 9:
            self.details['format'] = f"{Colors.GREEN}‚úÖ Formato correcto ({points}/10 pts){Colors.ENDC}"
        elif points >= 6:
            self.details['format'] = f"{Colors.YELLOW}‚ö†Ô∏è  Formato parcial ({points}/10 pts){Colors.ENDC}"
        else:
            self.details['format'] = f"{Colors.RED}‚ùå Formato con errores ({points}/10 pts): {', '.join(issues)}{Colors.ENDC}"
        
        return points

# ==================== FUNCI√ìN PRINCIPAL ====================

def main():
    print(f"{Colors.BOLD}{Colors.CYAN}")
    print("""
    ‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
    ‚ïë                                                                           ‚ïë
    ‚ïë           ü•ä PRUEBA 3: EXTRACCI√ìN DE DATOS ESTRUCTURADOS ü•ä              ‚ïë
    ‚ïë                                                                           ‚ïë
    ‚ïë                3 Emails - 3 Modelos - 1 Ganador                          ‚ïë
    ‚ïë                                                                           ‚ïë
    ‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù
    """)
    print(Colors.ENDC)
    
    # ==================== DATOS ESPERADOS ====================
    
    print_section("üìß DATOS ESPERADOS (GROUND TRUTH)")
    
    print(f"{Colors.BOLD}Email 1 - Mar√≠a Garc√≠a (TechSolutions SL):{Colors.ENDC}")
    print(f"  ‚Ä¢ 3 l√≠neas de productos")
    print(f"  ‚Ä¢ Presupuesto m√°x: 8,500 EUR")
    print(f"  ‚Ä¢ Fecha: 30 marzo ‚Üí {Colors.RED}URGENTE (15 d√≠as){Colors.ENDC}")
    print()
    
    print(f"{Colors.BOLD}Email 2 - Juan L√≥pez (InnovaStart):{Colors.ENDC}")
    print(f"  ‚Ä¢ 3 l√≠neas de productos + CIF")
    print(f"  ‚Ä¢ Presupuesto: 12,000-15,000 EUR")
    print(f"  ‚Ä¢ Fecha: 15 abril ‚Üí {Colors.RED}URGENTE (28 d√≠as){Colors.ENDC}")
    print()
    
    print(f"{Colors.BOLD}Email 3 - Ana Mart√≠nez (Comercial del Norte SL):{Colors.ENDC}")
    print(f"  ‚Ä¢ 4 l√≠neas de productos")
    print(f"  ‚Ä¢ Presupuesto: 6,000 EUR")
    print(f"  ‚Ä¢ Fecha: Mayo ‚Üí {Colors.YELLOW}BAJA (no urgente){Colors.ENDC}")
    print()
    
    print(f"{Colors.BOLD}TOTALES ESPERADOS:{Colors.ENDC}")
    print(f"  ‚Ä¢ Pedidos: 3")
    print(f"  ‚Ä¢ L√≠neas de productos: 9-10")
    print(f"  ‚Ä¢ Presupuesto min: 26,500 EUR (8500 + 12000 + 6000)")
    print(f"  ‚Ä¢ Presupuesto max: 29,500 EUR (8500 + 15000 + 6000)")
    print(f"  ‚Ä¢ Pedidos urgentes: 2")
    
    # ==================== EVALUACI√ìN ====================
    
    print_section("ü§ñ EVALUACI√ìN AUTOM√ÅTICA")
    
    evaluators = [
        DataExtractionEvaluator('gpt4o', GPT4O_JSON),
        DataExtractionEvaluator('qwen', QWEN_JSON),
        DataExtractionEvaluator('mixtral', MIXTRAL_JSON)
    ]
    
    results = []
    for evaluator in evaluators:
        result = evaluator.evaluate_all()
        results.append(result)
        
        info = MODEL_INFO[result['model_key']]
        print(f"\n{Colors.BOLD}{info['color']}{result['model_name']}:{Colors.ENDC}")
        print(f"  Hardware: {info['params']} | {info['vram']} | ~{info['speed']}s")
        print(f"  {Colors.BOLD}Puntuaci√≥n: {result['score']}/100{Colors.ENDC}")
        print()
        for key, value in result['details'].items():
            print(f"  {value}")
    
    # Ordenar por puntuaci√≥n
    results.sort(key=lambda x: x['score'], reverse=True)
    
    # Detectar empates (‚â§3 puntos)
    top_score = results[0]['score']
    tied_models = [r for r in results if top_score - r['score'] <= 3]
    
    # ==================== RESULTADOS FINALES ====================
    
    print_section("üèÜ RESULTADOS FINALES")
    
    if len(tied_models) > 1:
        print(f"{Colors.BOLD}{Colors.YELLOW}‚öñÔ∏è  EMPATE T√âCNICO (diferencia ‚â§3 pts){Colors.ENDC}\n")
    
    print(f"{Colors.BOLD}{'Pos':<6} {'Modelo':<25} {'Score':<15} {'Velocidad':<12}{Colors.ENDC}")
    print("‚îÄ" * 70)
    
    for i, result in enumerate(results):
        medal = "ü•á" if i == 0 else "ü•à" if i == 1 else "ü•â"
        if i < len(tied_models):
            medal = "üèÜ"
        
        info = MODEL_INFO[result['model_key']]
        print(f"{medal} {i+1}¬∫  {result['model_name']:<25} {result['score']}/100{'':7} ~{info['speed']}s")
    
    # ==================== AN√ÅLISIS DETALLADO ====================
    
    print_section("üîç AN√ÅLISIS DETALLADO")
    
    print(f"{Colors.CYAN}Diferencias clave encontradas:{Colors.ENDC}\n")
    
    # GPT-4o
    gpt4o_res = results[[r['model_key'] for r in results].index('gpt4o')]
    print(f"{Colors.RED}GPT-4o:{Colors.ENDC}")
    print(f"  ‚úÖ Estructura perfecta, 3 pedidos completos")
    print(f"  ‚úÖ Todas las cantidades extra√≠das")
    print(f"  ‚úÖ Urgencia clasificada correctamente")
    print(f"  ‚ùå Error en presupuesto_total_min: 18,000 (deber√≠a ser 26,500)")
    print(f"      ‚Üí No sum√≥ el email 1 (8,500) ni el 3 (6,000)")
    print()
    
    # Qwen
    qwen_res = results[[r['model_key'] for r in results].index('qwen')]
    print(f"{Colors.GREEN}Qwen 2.5 14B:{Colors.ENDC}")
    print(f"  ‚úÖ Estructura perfecta, 3 pedidos completos")
    print(f"  ‚úÖ Detalles extra: '(CTO)' en nombre cliente")
    print(f"  ‚ö†Ô∏è  Email 2: urgencia 'media' (deber√≠a ser 'alta')")
    print(f"  ‚ùå Cantidad=null en teclados/ratones")
    print(f"  ‚ùå Error en presupuesto_total_min: 12,000 (solo email 2)")
    print()
    
    # Mixtral
    mixtral_res = results[[r['model_key'] for r in results].index('mixtral')]
    print(f"{Colors.YELLOW}Mixtral 8x7B:{Colors.ENDC}")
    print(f"  ‚úÖ Estructura perfecta, 3 pedidos completos")
    print(f"  ‚úÖ Interpret√≥ bien teclados+ratones: 16 items (8 de cada)")
    print(f"  ‚ö†Ô∏è  Email 2: urgencia 'media' (deber√≠a ser 'alta')")
    print(f"  ‚ùå Presupuesto email 3: max=null (deber√≠a ser 6000)")
    print(f"  ‚ùå M√∫ltiples errores en resumen:")
    print(f"      ‚Üí presupuesto_total_min: 21,000 ‚â† 26,500")
    print(f"      ‚Üí presupuesto_total_max: 23,500 ‚â† 29,500")
    print(f"      ‚Üí pedidos_urgentes: 1 ‚â† 2")
    
    # ==================== VEREDICTO FINAL ====================
    
    print_section("üéØ VEREDICTO FINAL")
    
    winner = results[0]
    winner_info = MODEL_INFO[winner['model_key']]
    
    if len(tied_models) > 1:
        print(f"{Colors.BOLD}{Colors.YELLOW}üèÜ EMPATE: {' & '.join([t['model_name'] for t in tied_models])}{Colors.ENDC}\n")
    else:
        print(f"{Colors.BOLD}{winner_info['color']}üèÜ GANADOR: {winner['model_name']} ({winner['score']}/100){Colors.ENDC}\n")
    
    print(f"{Colors.CYAN}Conclusiones:{Colors.ENDC}\n")
    
    print(f"1. {Colors.BOLD}Extracci√≥n b√°sica:{Colors.ENDC} Los 3 modelos extrajeron estructura perfecta")
    print(f"   ‚Üí Todos capturaron los 3 emails con campos completos")
    print()
    
    print(f"2. {Colors.BOLD}Clasificaci√≥n de urgencia:{Colors.ENDC}")
    print(f"   ‚Üí GPT-4o: Perfecto (alta, alta, baja)")
    print(f"   ‚Üí Qwen & Mixtral: Clasificaron email 2 como 'media'")
    print(f"     (Debatible: 28 d√≠as podr√≠a ser media o alta)")
    print()
    
    print(f"3. {Colors.BOLD}C√°lculos de resumen:{Colors.ENDC}")
    print(f"   ‚Üí TODOS fallaron en al menos un c√°lculo")
    print(f"   ‚Üí GPT-4o: Error grave en presupuesto_min")
    print(f"   ‚Üí Qwen: Error similar en presupuesto_min")
    print(f"   ‚Üí Mixtral: M√∫ltiples errores en resumen")
    print()
    
    print(f"{Colors.BOLD}{Colors.GREEN}üí° LECCI√ìN CLAVE:{Colors.ENDC}")
    print(f"  Extracci√≥n de campos: {Colors.GREEN}Los 3 modelos excelentes{Colors.ENDC}")
    print(f"  C√°lculos matem√°ticos: {Colors.RED}Todos tienen problemas{Colors.ENDC}")
    print()
    print(f"  ‚Üí Para extracci√≥n pura: Modelos locales funcionan perfecto")
    print(f"  ‚Üí Para validaci√≥n de sumas: Mejor a√±adir validaci√≥n post-proceso")
    
    # ==================== COST-BENEFIT ====================
    
    print_section("üí∞ AN√ÅLISIS COSTO-BENEFICIO")
    
    print(f"{Colors.BOLD}Escenario: 10,000 extracciones/mes{Colors.ENDC}\n")
    
    for r in results:
        info = MODEL_INFO[r['model_key']]
        monthly_cost = info['cost_per_1k'] * 10
        yearly_cost = monthly_cost * 12
        
        print(f"{info['color']}{r['model_name']}:{Colors.ENDC}")
        print(f"  Score: {r['score']}/100")
        print(f"  Costo: ${monthly_cost:.2f}/mes ‚Üí ${yearly_cost:.2f}/a√±o")
        if r['score'] > 0:
            print(f"  Costo por punto: ${yearly_cost/r['score']:.2f}")
        print()
    
    print(f"{Colors.CYAN}Recomendaci√≥n:{Colors.ENDC}")
    if winner['model_key'] != 'gpt4o':
        print(f"  Para extracci√≥n de emails/documentos: {winner_info['color']}{winner['model_name']}{Colors.ENDC}")
        print(f"  Calidad similar, completamente gratis")
        print(f"  Ahorro: $600/a√±o (10K extracciones/mes)")
    else:
        print(f"  Para m√°xima precisi√≥n: {Colors.RED}GPT-4o{Colors.ENDC}")
        print(f"  Para ahorro: {Colors.GREEN}Qwen 14B{Colors.ENDC} (muy similar, $0)")
    
    print()

if __name__ == "__main__":
    main()
```

```python
#Prueba 4
"""
An√°lisis exhaustivo: Prueba 4 - Servicio al Cliente
GPT-4o vs Llama 3.1 8B vs Mistral 7B
Con datos reales de ejecuci√≥n
"""

import json
import re

# Colores
class Colors:
    HEADER = '\033[95m'
    BLUE = '\033[94m'
    CYAN = '\033[96m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

def print_section(title: str):
    print(f"\n{Colors.BOLD}{Colors.HEADER}{'='*90}{Colors.ENDC}")
    print(f"{Colors.BOLD}{Colors.HEADER}{title:^90}{Colors.ENDC}")
    print(f"{Colors.BOLD}{Colors.HEADER}{'='*90}{Colors.ENDC}\n")

# ==================== OUTPUTS REALES ====================

GPT4O_JSON = """{
  "asunto": "Re: Problema con mi pedido #TR-2847",
  "cuerpo": "Hola Carlos,\\n\\nGracias por contactarte con nosotros. Lamentamos sinceramente la demora en el env√≠o de tu pedido #TR-2847. Comprendemos tu frustraci√≥n y estamos aqu√≠ para ayudarte. \\n\\nLa causa del retraso fue un imprevisto con el stock, que ya hemos solucionado. Nos complace informarte que tu Lenovo ThinkPad X1 Carbon Gen 11 est√° previsto para ser enviado ma√±ana a trav√©s de DHL Express, con una entrega estimada en 1-2 d√≠as, por lo cual deber√≠a llegarte antes del lunes.\\n\\nPara seguir el estado de tu pedido, puedes utilizar el siguiente enlace: https://techstore.com/tracking/TR-2847\\n\\nPara compensar las molestias causadas, nos gustar√≠a ofrecerte un descuento del 10% en tu pr√≥xima compra. \\n\\nAgradecemos tu comprensi√≥n y paciencia. Si tienes alguna otra pregunta o necesitas m√°s ayuda, no dudes en contactarnos.\\n\\nSaludos cordiales,\\nEquipo de Atenci√≥n al Cliente - TechStore",
  "tono": "emp√°tico/profesional/c√°lido",
  "compensacion_ofrecida": "10% de descuento en pr√≥xima compra",
  "problema_resuelto": true,
  "tiempo_respuesta_estimado": "inmediato"
}"""

LLAMA_JSON = """{
  "asunto": "Re: Problema con mi pedido #TR-2847",
  "cuerpo": "Hola Carlos,\\n\\nQueremos disculparnos por la demora en el env√≠o de tu port√°til Lenovo ThinkPad X1 Carbon Gen 11. Comprendemos que necesitas este equipo urgentemente para tu nuevo trabajo.\\n\\nLa situaci√≥n es que, debido a un stock temporalmente agotado, hubo un retraso en nuestro almac√©n. Sin embargo, hemos recibido la reposici√≥n ayer y estamos preparando el env√≠o inmediatamente. El estado de tu pedido est√° actualizado a "En preparaci√≥n en almac√©n".\\n\\nRecibir√°s una actualizaci√≥n del transportista DHL Express (1-2 d√≠as de entrega) tan pronto como se env√≠e. Puedes rastrear tu pedido en: https://techstore.com/tracking/TR-2847.\\n\\nComo gesto de buena voluntad, te ofrecemos un descuento del 10% en tu pr√≥xima compra con nosotros.\\n\\nAgradecemos tu paciencia y apoyo.\\n\\nAtentamente,\\nEquipo de Atenci√≥n al Cliente - TechStore",
  "tono": "emp√°tico/profesional/c√°lido",
  "compensacion_ofrecida": "Descuento del 10% en la pr√≥xima compra",
  "problema_resuelto": false,
  "tiempo_respuesta_estimado": "inmediato"
}"""

MISTRAL_JSON = """{
  "asunto": "Re: [Problema con mi pedido #TR-2847]",
  "cuerpo": "Estimado Carlos Ruiz,\\n\\nLe informamos que su pedido con n√∫mero de referencia #TR-2847 est√° en proceso de env√≠o. Hemos tenido un retraso temporal debido a un agotamiento de stock, pero hemos recibido la reposici√≥n del producto Lenovo ThinkPad X1 Carbon Gen 11 ayer.\\n\\nEl pedido se estar√° enviando ma√±ana (24 horas), y llegar√° entre 1-2 d√≠as despu√©s de su env√≠o. La entrega ser√° realizada por DHL Express.\\n\\nPor su ayuda inesperada, nos permiteme ofrecerle una descuento de un 10% en su siguiente pedido.\\n\\nLe agradecemos su comprensi√≥n y le recordamos que puede rastrear su pedido en la siguiente URL: https://techstore.com/tracking/TR-2847.\\n\\nSaludos cordiales,\\nEquipo de Atenci√≥n al Cliente - TechStore",
  "tono": "emp√°tico y profesional",
  "compensacion_ofrecida": "descuento del 10% en el siguiente pedido",
  "problema_resuelto": true,
  "tiempo_respuesta_estimado": "inmediato"
}"""

# ==================== INFORMACI√ìN DE MODELOS ====================

MODEL_INFO = {
    'gpt4o': {
        'name': 'GPT-4o',
        'params': '1-2 Trillones',
        'cost_per_1k': 0.05,
        'speed': 4.5,
        'color': Colors.RED
    },
    'llama': {
        'name': 'Llama 3.1 8B',
        'params': '8 Billones',
        'cost_per_1k': 0.00,
        'speed': 3.5,
        'color': Colors.GREEN
    },
    'mistral': {
        'name': 'Mistral 7B',
        'params': '7 Billones',
        'cost_per_1k': 0.00,
        'speed': 3.0,
        'color': Colors.YELLOW
    }
}

# ==================== EVALUADOR ====================

class CustomerServiceEvaluator:
    def __init__(self, model_key: str, json_data: str):
        self.model_key = model_key
        self.info = MODEL_INFO[model_key]
        try:
            self.data = json.loads(json_data)
        except Exception as e:
            print(f"{Colors.RED}Error parseando {model_key}: {e}{Colors.ENDC}")
            self.data = {}
        self.score = 0
        self.details = {}
        
    def evaluate_all(self):
        """Ejecuta todas las evaluaciones"""
        
        # 1. Estructura del mensaje (15 pts)
        structure_score = self.evaluate_structure()
        
        # 2. Empat√≠a y tono (25 pts)
        empathy_score = self.evaluate_empathy()
        
        # 3. Resoluci√≥n del problema (30 pts)
        resolution_score = self.evaluate_resolution()
        
        # 4. Compensaci√≥n apropiada (15 pts)
        compensation_score = self.evaluate_compensation()
        
        # 5. Calidad del lenguaje (15 pts)
        language_score = self.evaluate_language()
        
        self.score = (structure_score + empathy_score + resolution_score + 
                     compensation_score + language_score)
        
        return {
            'model_key': self.model_key,
            'model_name': self.info['name'],
            'score': self.score,
            'details': self.details,
            'data': self.data
        }
    
    def evaluate_structure(self) -> int:
        """Eval√∫a la estructura del mensaje"""
        points = 0
        issues = []
        
        asunto = self.data.get('asunto', '')
        cuerpo = self.data.get('cuerpo', '')
        
        # Asunto apropiado (5 pts)
        if 'Re:' in asunto and '#TR-2847' in asunto:
            if '[' not in asunto and ']' not in asunto:
                points += 5
            else:
                points += 3
                issues.append("corchetes innecesarios en asunto")
        else:
            issues.append("asunto incompleto")
        
        # Saludo personalizado (3 pts)
        if 'Carlos' in cuerpo:
            if 'Hola Carlos' in cuerpo:
                points += 3
            elif 'Estimado Carlos' in cuerpo:
                points += 2
        else:
            issues.append("sin nombre del cliente")
        
        # Despedida profesional (4 pts)
        if 'Saludos' in cuerpo or 'Atentamente' in cuerpo:
            points += 4
        
        # Firma (3 pts)
        if 'Equipo de Atenci√≥n al Cliente' in cuerpo or 'TechStore' in cuerpo:
            points += 3
        
        if points >= 13:
            self.details['structure'] = f"{Colors.GREEN}‚úÖ Estructura excelente ({points}/15){Colors.ENDC}"
        elif points >= 10:
            self.details['structure'] = f"{Colors.YELLOW}‚ö†Ô∏è  Estructura buena ({points}/15): {', '.join(issues)}{Colors.ENDC}"
        else:
            self.details['structure'] = f"{Colors.RED}‚ùå Estructura mejorable ({points}/15){Colors.ENDC}"
        
        return points
    
    def evaluate_empathy(self) -> int:
        """Eval√∫a empat√≠a y tono"""
        points = 0
        issues = []
        
        cuerpo = self.data.get('cuerpo', '').lower()
        tono = self.data.get('tono', '').lower()
        
        # Disculpa expl√≠cita (7 pts)
        if 'lamentamos' in cuerpo or 'disculp' in cuerpo:
            points += 7
        else:
            issues.append("sin disculpa expl√≠cita")
        
        # Reconoce frustraci√≥n/situaci√≥n del cliente (7 pts)
        if 'comprendemos' in cuerpo or 'entendemos' in cuerpo:
            points += 7
            # Bonus si menciona contexto espec√≠fico
            if 'trabajo' in cuerpo or 'urgente' in cuerpo:
                points += 2
        else:
            issues.append("no reconoce situaci√≥n del cliente")
        
        # Tono apropiado (6 pts)
        if 'emp√°tico' in tono and 'profesional' in tono:
            points += 6
        elif 'emp√°tico' in tono or 'profesional' in tono:
            points += 4
        
        # Lenguaje c√°lido vs rob√≥tico (5 pts)
        warm_phrases = ['gracias por', 'nos complace', 'estamos aqu√≠', 'queremos', 'gesto de buena voluntad']
        cold_phrases = ['le informamos', 'se le notifica']
        
        warm_count = sum(1 for phrase in warm_phrases if phrase in cuerpo)
        cold_count = sum(1 for phrase in cold_phrases if phrase in cuerpo)
        
        if warm_count >= 2:
            points += 5
        elif warm_count >= 1:
            points += 3
        elif cold_count >= 1:
            points += 1
            issues.append("lenguaje demasiado formal/fr√≠o")
        
        if points >= 23:
            self.details['empathy'] = f"{Colors.GREEN}‚úÖ Empat√≠a excelente ({points}/25){Colors.ENDC}"
        elif points >= 18:
            self.details['empathy'] = f"{Colors.YELLOW}‚ö†Ô∏è  Empat√≠a buena ({points}/25): {', '.join(issues)}{Colors.ENDC}"
        else:
            self.details['empathy'] = f"{Colors.RED}‚ùå Empat√≠a insuficiente ({points}/25){Colors.ENDC}"
        
        return min(25, points)
    
    def evaluate_resolution(self) -> int:
        """Eval√∫a la resoluci√≥n del problema"""
        points = 0
        issues = []
        
        cuerpo = self.data.get('cuerpo', '')
        problema_resuelto = self.data.get('problema_resuelto', False)
        
        # Explica la causa (8 pts)
        if 'stock' in cuerpo.lower():
            points += 8
        else:
            issues.append("no explica causa")
        
        # Ofrece soluci√≥n concreta (10 pts)
        if 'ma√±ana' in cuerpo.lower() or 'enviado' in cuerpo.lower():
            if problema_resuelto:
                points += 10
            else:
                points += 7
                issues.append("soluci√≥n no confirmada definitivamente")
        else:
            issues.append("sin timeframe concreto")
        
        # Da informaci√≥n de seguimiento (7 pts)
        if 'tracking' in cuerpo.lower() or 'rastrear' in cuerpo.lower():
            if 'https://techstore.com/tracking/TR-2847' in cuerpo:
                points += 7
            else:
                points += 5
        else:
            issues.append("sin info de seguimiento")
        
        # Menciona transportista y tiempos (5 pts)
        if 'DHL' in cuerpo and ('1-2 d√≠as' in cuerpo or '24 horas' in cuerpo):
            points += 5
        elif 'DHL' in cuerpo or '1-2 d√≠as' in cuerpo:
            points += 3
        
        if points >= 27:
            self.details['resolution'] = f"{Colors.GREEN}‚úÖ Resoluci√≥n excelente ({points}/30){Colors.ENDC}"
        elif points >= 22:
            self.details['resolution'] = f"{Colors.YELLOW}‚ö†Ô∏è  Resoluci√≥n buena ({points}/30): {', '.join(issues)}{Colors.ENDC}"
        else:
            self.details['resolution'] = f"{Colors.RED}‚ùå Resoluci√≥n insuficiente ({points}/30){Colors.ENDC}"
        
        return points
    
    def evaluate_compensation(self) -> int:
        """Eval√∫a la compensaci√≥n ofrecida"""
        points = 0
        
        comp = self.data.get('compensacion_ofrecida', '').lower()
        cuerpo = self.data.get('cuerpo', '').lower()
        
        # Ofrece compensaci√≥n (8 pts)
        if comp and ('descuento' in comp or '10%' in comp):
            points += 8
        
        # Es espec√≠fica (4 pts)
        if '10%' in comp or '10%' in cuerpo:
            points += 4
        
        # Est√° en el cuerpo del mensaje (3 pts)
        if 'descuento' in cuerpo and ('10%' in cuerpo or 'pr√≥xima compra' in cuerpo):
            points += 3
        
        if points >= 13:
            self.details['compensation'] = f"{Colors.GREEN}‚úÖ Compensaci√≥n clara ({points}/15){Colors.ENDC}"
        elif points >= 10:
            self.details['compensation'] = f"{Colors.YELLOW}‚ö†Ô∏è  Compensaci√≥n presente ({points}/15){Colors.ENDC}"
        else:
            self.details['compensation'] = f"{Colors.RED}‚ùå Compensaci√≥n insuficiente ({points}/15){Colors.ENDC}"
        
        return points
    
    def evaluate_language(self) -> int:
        """Eval√∫a calidad del lenguaje"""
        points = 0
        issues = []
        
        cuerpo = self.data.get('cuerpo', '')
        
        # Sin errores gramaticales obvios (8 pts)
        errores = []
        if 'nos permiteme' in cuerpo.lower():
            errores.append("'nos permiteme' ‚Üí 'nos perm√≠teme' o 'perm√≠tenos'")
        if 'una descuento' in cuerpo.lower():
            errores.append("'una descuento' ‚Üí 'un descuento'")
        if 'por su ayuda inesperada' in cuerpo.lower():
            errores.append("'por su ayuda inesperada' sin sentido (¬ø'por las molestias'?)")
        
        if not errores:
            points += 8
        elif len(errores) == 1:
            points += 5
            issues.extend(errores)
        else:
            points += 2
            issues.extend(errores)
        
        # Lenguaje natural y fluido (4 pts)
        if len(cuerpo) > 200:  # Respuesta completa
            points += 4
        
        # Usa nombre del cliente (3 pts)
        if 'Carlos' in cuerpo:
            points += 3
        
        if points >= 13:
            self.details['language'] = f"{Colors.GREEN}‚úÖ Lenguaje excelente ({points}/15){Colors.ENDC}"
        elif points >= 10:
            self.details['language'] = f"{Colors.YELLOW}‚ö†Ô∏è  Lenguaje bueno ({points}/15): {', '.join(issues)}{Colors.ENDC}"
        else:
            self.details['language'] = f"{Colors.RED}‚ùå Lenguaje con errores ({points}/15): {', '.join(issues)}{Colors.ENDC}"
        
        return points

# ==================== FUNCI√ìN PRINCIPAL ====================

def main():
    print(f"{Colors.BOLD}{Colors.CYAN}")
    print("""
    ‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
    ‚ïë                                                                           ‚ïë
    ‚ïë              ü•ä PRUEBA 4: SERVICIO AL CLIENTE ü•ä                         ‚ïë
    ‚ïë                                                                           ‚ïë
    ‚ïë           1 Queja de Cliente - 3 Modelos - 1 Ganador                     ‚ïë
    ‚ïë                                                                           ‚ïë
    ‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù
    """)
    print(Colors.ENDC)
    
    # ==================== CONTEXTO ====================
    
    print_section("üìß CONTEXTO DE LA QUEJA")
    
    print(f"{Colors.BOLD}Email del cliente (Carlos Ruiz):{Colors.ENDC}\n")
    print(f"  De: carlos.ruiz@email.com")
    print(f"  Asunto: Problema con mi pedido #TR-2847")
    print(f"  Fecha: 20 de marzo de 2024\n")
    print(f"  {Colors.YELLOW}Contenido:{Colors.ENDC}")
    print(f"  'Hola,")
    print(f"  Hace m√°s de 2 semanas que hice un pedido de un Lenovo ThinkPad")
    print(f"  X1 Carbon Gen 11 (pedido #TR-2847) y todav√≠a no lo he recibido.")
    print(f"  En la web dice 'En proceso' pero no he recibido ninguna")
    print(f"  actualizaci√≥n. Necesito el port√°til urgentemente para mi nuevo")
    print(f"  trabajo que empieza el lunes. ¬øQu√© est√° pasando?'")
    print()
    print(f"{Colors.BOLD}Expectativas:{Colors.ENDC}")
    print(f"  ‚Ä¢ {Colors.GREEN}Respuesta emp√°tica{Colors.ENDC} (reconocer frustraci√≥n)")
    print(f"  ‚Ä¢ {Colors.GREEN}Explicar causa{Colors.ENDC} del retraso")
    print(f"  ‚Ä¢ {Colors.GREEN}Soluci√≥n concreta{Colors.ENDC} con timeframes")
    print(f"  ‚Ä¢ {Colors.GREEN}Compensaci√≥n{Colors.ENDC} por las molestias")
    print(f"  ‚Ä¢ {Colors.GREEN}Tono profesional pero c√°lido{Colors.ENDC}")
    
    # ==================== EVALUACI√ìN ====================
    
    print_section("ü§ñ EVALUACI√ìN AUTOM√ÅTICA")
    
    evaluators = [
        CustomerServiceEvaluator('gpt4o', GPT4O_JSON),
        CustomerServiceEvaluator('llama', LLAMA_JSON),
        CustomerServiceEvaluator('mistral', MISTRAL_JSON)
    ]
    
    results = []
    for evaluator in evaluators:
        result = evaluator.evaluate_all()
        results.append(result)
        
        info = MODEL_INFO[result['model_key']]
        print(f"\n{Colors.BOLD}{info['color']}{result['model_name']}:{Colors.ENDC}")
        print(f"  {Colors.BOLD}Puntuaci√≥n: {result['score']}/100{Colors.ENDC}")
        print(f"  Velocidad: ~{info['speed']}s | Costo: ${info['cost_per_1k']:.2f}/1K")
        print()
        for key, value in result['details'].items():
            print(f"  {value}")
    
    # Ordenar por puntuaci√≥n
    results.sort(key=lambda x: x['score'], reverse=True)
    
    # Detectar empates (‚â§3 puntos)
    top_score = results[0]['score']
    tied_models = [r for r in results if top_score - r['score'] <= 3]
    
    # ==================== RESULTADOS FINALES ====================
    
    print_section("üèÜ RESULTADOS FINALES")
    
    if len(tied_models) > 1:
        print(f"{Colors.BOLD}{Colors.YELLOW}‚öñÔ∏è  EMPATE T√âCNICO (diferencia ‚â§3 pts){Colors.ENDC}\n")
    
    print(f"{Colors.BOLD}{'Pos':<6} {'Modelo':<25} {'Score':<15} {'Velocidad':<12}{Colors.ENDC}")
    print("‚îÄ" * 70)
    
    for i, result in enumerate(results):
        medal = "ü•á" if i == 0 else "ü•à" if i == 1 else "ü•â"
        if i < len(tied_models):
            medal = "üèÜ"
        
        info = MODEL_INFO[result['model_key']]
        print(f"{medal} {i+1}¬∫  {result['model_name']:<25} {result['score']}/100{'':7} ~{info['speed']}s")
    
    # ==================== AN√ÅLISIS DETALLADO ====================
    
    print_section("üîç AN√ÅLISIS DETALLADO DE RESPUESTAS")
    
    # GPT-4o
    gpt4o_res = next(r for r in results if r['model_key'] == 'gpt4o')
    print(f"{Colors.RED}{Colors.BOLD}GPT-4o:{Colors.ENDC}")
    print(f"  {Colors.GREEN}Fortalezas:{Colors.ENDC}")
    print(f"    ‚Ä¢ Empat√≠a excepcional: 'Lamentamos sinceramente', 'Comprendemos tu frustraci√≥n'")
    print(f"    ‚Ä¢ Informaci√≥n completa: causa, soluci√≥n, timeframe, tracking, compensaci√≥n")
    print(f"    ‚Ä¢ Tono perfecto: profesional pero c√°lido")
    print(f"    ‚Ä¢ Estructura impecable: saludo, cuerpo, despedida")
    print(f"  {Colors.YELLOW}√Åreas de mejora:{Colors.ENDC}")
    print(f"    ‚Ä¢ Quiz√°s demasiado 'perfecto'/gen√©rico")
    print()
    
    # Llama
    llama_res = next(r for r in results if r['model_key'] == 'llama')
    print(f"{Colors.GREEN}{Colors.BOLD}Llama 3.1 8B:{Colors.ENDC}")
    print(f"  {Colors.GREEN}Fortalezas:{Colors.ENDC}")
    print(f"    ‚Ä¢ Personalizaci√≥n: menciona 'nuevo trabajo' del cliente")
    print(f"    ‚Ä¢ Empat√≠a natural: 'Queremos disculparnos'")
    print(f"    ‚Ä¢ Contexto adicional: 'En preparaci√≥n en almac√©n'")
    print(f"  {Colors.YELLOW}√Åreas de mejora:{Colors.ENDC}")
    print(f"    ‚Ä¢ problema_resuelto: false (ambig√ºedad en si est√° enviado)")
    print(f"    ‚Ä¢ No confirma env√≠o definitivo, solo 'preparando'")
    print()
    
    # Mistral
    mistral_res = next(r for r in results if r['model_key'] == 'mistral')
    print(f"{Colors.YELLOW}{Colors.BOLD}Mistral 7B:{Colors.ENDC}")
    print(f"  {Colors.GREEN}Fortalezas:{Colors.ENDC}")
    print(f"    ‚Ä¢ Informaci√≥n clara y concisa")
    print(f"    ‚Ä¢ Timeframes espec√≠ficos: 'ma√±ana (24 horas)', '1-2 d√≠as'")
    print(f"    ‚Ä¢ Estructura correcta")
    print(f"  {Colors.RED}Debilidades:{Colors.ENDC}")
    print(f"    ‚Ä¢ ERROR GRAMATICAL: 'Por su ayuda inesperada' (sin sentido)")
    print(f"    ‚Ä¢ ERROR: 'nos permiteme' ‚Üí 'nos perm√≠teme' o 'perm√≠tenos'")
    print(f"    ‚Ä¢ ERROR: 'una descuento' ‚Üí 'un descuento'")
    print(f"    ‚Ä¢ Tono m√°s fr√≠o/formal: 'Le informamos', 'Estimado'")
    print(f"    ‚Ä¢ Corchetes innecesarios en asunto: [Problema...]")
    
    # ==================== VEREDICTO FINAL ====================
    
    print_section("üéØ VEREDICTO FINAL")
    
    winner = results[0]
    winner_info = MODEL_INFO[winner['model_key']]
    
    if len(tied_models) > 1:
        print(f"{Colors.BOLD}{Colors.YELLOW}üèÜ EMPATE: {' & '.join([t['model_name'] for t in tied_models])}{Colors.ENDC}\n")
    else:
        print(f"{Colors.BOLD}{winner_info['color']}üèÜ GANADOR: {winner['model_name']} ({winner['score']}/100){Colors.ENDC}\n")
    
    print(f"{Colors.CYAN}Puntuaciones finales:{Colors.ENDC}")
    for r in results:
        info = MODEL_INFO[r['model_key']]
        print(f"  {info['color']}{r['model_name']}: {r['score']}/100{Colors.ENDC}")
    
    print(f"\n{Colors.CYAN}An√°lisis:{Colors.ENDC}\n")
    
    print(f"1. {Colors.BOLD}Calidad general:{Colors.ENDC}")
    print(f"   ‚Üí GPT-4o: Respuesta de manual, perfecta t√©cnicamente")
    print(f"   ‚Üí Llama: Respuesta natural y personalizada")
    print(f"   ‚Üí Mistral: Respuesta funcional pero con errores")
    print()
    
    print(f"2. {Colors.BOLD}Empat√≠a y tono:{Colors.ENDC}")
    print(f"   ‚Üí GPT-4o y Llama: Excelente, c√°lidos y profesionales")
    print(f"   ‚Üí Mistral: M√°s formal/fr√≠o, menos cercano")
    print()
    
    print(f"3. {Colors.BOLD}Errores cr√≠ticos:{Colors.ENDC}")
    print(f"   ‚Üí GPT-4o: Ninguno")
    print(f"   ‚Üí Llama: Ambig√ºedad (no confirma env√≠o definitivo)")
    print(f"   ‚Üí Mistral: M√∫ltiples errores gramaticales graves")
    print()
    
    print(f"{Colors.BOLD}{Colors.GREEN}üí° CONCLUSI√ìN:{Colors.ENDC}")
    
    if winner['model_key'] == 'gpt4o':
        print(f"  Para servicio al cliente cr√≠tico:")
        print(f"  {Colors.RED}GPT-4o es superior.{Colors.ENDC}")
        print(f"  Lenguaje impecable, empat√≠a perfecta, sin errores.")
        print()
        print(f"  {Colors.GREEN}Llama 3.1 8B{Colors.ENDC} es una alternativa muy v√°lida:")
        print(f"  M√°s personalizado, gratis, solo 7 puntos menos.")
        print()
        print(f"  {Colors.YELLOW}Mistral 7B{Colors.ENDC} necesita mejora:")
        print(f"  Errores gramaticales inaceptables en atenci√≥n al cliente.")
    
    # ==================== COST-BENEFIT ====================
    
    print_section("üí∞ AN√ÅLISIS COSTO-BENEFICIO")
    
    print(f"{Colors.BOLD}Escenario: 10,000 respuestas/mes{Colors.ENDC}\n")
    
    for r in results:
        info = MODEL_INFO[r['model_key']]
        monthly_cost = info['cost_per_1k'] * 10
        yearly_cost = monthly_cost * 12
        
        print(f"{info['color']}{r['model_name']}:{Colors.ENDC}")
        print(f"  Score: {r['score']}/100")
        print(f"  Costo: ${monthly_cost:.2f}/mes ‚Üí ${yearly_cost:.2f}/a√±o")
        if r['score'] > 0:
            print(f"  Costo por punto: ${yearly_cost/r['score']:.2f}")
        print()
    
    print(f"{Colors.CYAN}Recomendaci√≥n:{Colors.ENDC}")
    print(f"  Para empresa grande con volumen alto:")
    print(f"    ‚Üí {Colors.GREEN}Llama 3.1 8B{Colors.ENDC} (gratis, 88/100)")
    print(f"    ‚Üí Ahorro: $600/a√±o vs GPT-4o")
    print()
    print(f"  Para casos cr√≠ticos o legal:")
    print(f"    ‚Üí {Colors.RED}GPT-4o{Colors.ENDC} (95/100, sin riesgo de errores)")
    
    print()

if __name__ == "__main__":
    main()
```
